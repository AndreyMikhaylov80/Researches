КРАТКОЕ ОПИСАНИЕ КОНЦЕПТА
=========================

Мы - поколение, осознавшее, что человек не временный пользователь, а
ответственный владелец планеты и собственной жизни. Наша цель – не
просто активность, а осмысленное бытие. Участие в движении не требует
специальных действий: достаточно **интерпретировать** свои повседневные
учебные и научные занятия как вклад в пять универсальных ценностей
(Энергия, Благо, Справедливость, Экология, Культура) и делиться этими
смыслами на Советах. Тем самым мы строим новую эпоху духовно развитого
человечества и проверяем себя: **«Правда ли мы то самое человечество,
которое может?»**.

Количественной основой движения служит **Индекс Цивилизационного
Развития (ICD)** – двухкомпонентная метрика, оценивающая гармоничность
личности (индивидуальный ICD) и её социальное влияние (индекс влияния).
Все деяния фиксируются в неизменном реестре с использованием блокчейна,
а каждый участник получает цифровой токен KIT (soulbound), хранящий его
достижения.

Организационная структура охватывает от первичных ячеек в вузах до
федерального и международного уровней, сочетая демократические принципы
с меритократией. Дорожная карта предусматривает поэтапное развитие от
идеи до всероссийского масштаба (TRL1–TRL6) с охватом 5% студентов в
каждом вузе.

Раздел 1. Миссия и философия движения

-   **Главный инструмент – Совет.** Пространство, где участники делятся

    ли мы то самое человечество, которое способно жить в гармонии,
    обеспечить каждому базовые блага и выйти к звёздам?»*

Раздел 2. Индекс Цивилизационного Развития (ICD)
------------------------------------------------

2.1. Двухкомпонентная модель
----------------------------

Для разделения понятий «личная гармоничность» и «социальное влияние»
введены два показателя:

-   **Индивидуальный ICD (гармоничность)** – характеризует
    сбалансированность развития участника по пяти лучам. Вычисляется на
    основе **исходных баллов**, не зависящих от популярности.

-   **Индекс влияния** – количественная мера воздействия деяний
    участника на других людей, определяемая через голосование.
    Используется для присвоения статусов (Мастер, Творец, Легенда) и
    выявления лидеров мнений.

2.2. Балльная система для деяний
--------------------------------

-   **Обычное деяние:** главный луч – 2 балла, дополнительные (до двух)
    – по 1 баллу.

-   **Прорывное деяние** (изобретения, открытия планетарного масштаба):
    Федеральный совет на основе внешней экспертизы может установить
    повышающий множитель (например, ×1000) к исходным баллам.

2.3. Базовый уровень
--------------------

Каждый новый участник при регистрации автоматически получает **по 1
баллу в каждый луч**. Это гарантирует:

-   отсутствие нулевых значений при расчёте среднего геометрического;

-   возможность новичкам иметь ненулевой ICD с самого начала.

2.4. Индивидуальный ICD (гармоничность)
---------------------------------------

Пусть $B_{E},B_{W},B_{J},B_{S},B_{C}$ – суммарные исходные баллы
участника (с учётом базовых).  
$M_{E},M_{W},M_{J},M_{S},M_{C}$ – текущие максимальные значения исходных
баллов среди всех участников движения.

Нормированные значения:

$$x_{l} = \frac{B_{l}}{M_{l}},l \in \{ E,W,J,S,C\}.
$$

Индивидуальный ICD:

$$\text{IC}D_{\text{личн}} = \sqrt[5]{x_{E} \cdot x_{W} \cdot x_{J} \cdot x_{S} \cdot x_{C}}.
$$

Среднее геометрическое выбрано потому, что оно резко падает при низких
значениях любого из сомножителей, стимулируя гармоничное развитие.

2.5. Учёт социального влияния (голосование)
-------------------------------------------

Каждый участник может проголосовать за деяние другого участника (один
раз за каждое деяние), если оно оказало на него положительное влияние.
Голоса фиксируются в реестре.

Для деяния с $V$ уникальными голосами:

$$k = 1 + \ln(1 + V).
$$

Логарифмическая функция предотвращает неограниченный рост и ослабляет
эффект «накруток».

**Взвешенные баллы** деяния по каждому лучу:

$$\text{Взвешенные\ баллы}_{l} = \text{Исходные\ баллы}_{l} \times k.
$$

2.6. Индекс влияния участника
-----------------------------

$$I_{\text{влияния}} = \sum_{\text{деяния}}^{}\left( E_{\text{взв}} + W_{\text{взв}} + J_{\text{взв}} + S_{\text{взв}} + C_{\text{взв}} \right).
$$

На основе индекса влияния автоматически присваиваются статусы (пороги
могут корректироваться):

-   **Мастер** – при $I > 1000$;

-   **Творец** – при $I > 5000$;

-   **Легенда** – при $I > 10000$.

2.7. Глобальный Уровень Цивилизации (УЦ)
----------------------------------------

УЦ определяется не по средним, а по **максимальным
достижениям** человечества в каждом луче (аналогия со шкалой Кардашёва).

Для каждого луча вводятся **абсолютные реперы** $L_{l}^{\text{abs}}$ –
значения, соответствующие гипотетическому уровню УЦ-5 (полное развитие).
Они фиксированы и не меняются.

Текущие максимумы исходных баллов $M_{l}$ нормируются:

$$y_{l} = \frac{M_{l}}{L_{l}^{\text{abs}}}.
$$

Глобальный УЦ:

$$\text{УЦ}(t) = \sqrt[5]{y_{E} \cdot y_{W} \cdot y_{J} \cdot y_{S} \cdot y_{C}}.
$$

Это значение автоматически обновляется при улучшении любого из
максимумов (например, после регистрации прорывного деяния).

|----------|------------------|-------------|-------------------------------------------------------------------|

Раздел 3. Учёт деяний и цифровой след
-------------------------------------

3.1. Принципы учёта
-------------------

-   **Неизменность** – записи о деяниях и голосах не могут быть
    подделаны.

-   **Прозрачность** – любой участник может проверить данные.

-   **Простота** – использование привычных инструментов (Google Таблицы,
    Telegram).

-   **Справедливость** – разделение объективной ценности и социального
    влияния.

3.2. Алгоритм регистрации деяния
--------------------------------

1.  **Инициация:** участник (или староста) заполняет форму в
    Telegram-боте или Google Forms, указывая описание, лучи и ссылку на
    подтверждение (фото, публикация).

2.  **Валидация:** староста группы проверяет заявку (3–5 дней). В
    спорных случаях – Совет первички.

3.  **Внесение в реестр исходных баллов:** утверждённые деяния заносятся
    в единую Google Таблицу «Деяния» с указанием исходных баллов (с
    учётом экспертного множителя, если применимо).

4.  **Базовые баллы:** при регистрации нового участника автоматически
    вносится запись о начислении 1 балла в каждый луч.

3.3. Особый порядок для прорывных деяний
----------------------------------------

Для деяний, которые могут претендовать на повышающий множитель, заявка
направляется в Федеральный совет с приложением экспертных заключений,
патентов, публикаций. Совет назначает внешнюю экспертизу и утверждает
множитель. Это необходимо для своевременного отражения объективного
вклада в глобальные максимумы $M_{l}$ и УЦ.

3.4. Сервис голосования
-----------------------

-   Голосование доступно всем участникам через Telegram-бота или
    веб-интерфейс.

-   Каждое деяние можно оценить только один раз.

-   Все голоса фиксируются в отдельной Google Таблице «Голоса» с
    указанием ID деяния, ID голосующего и времени.

-   Периодически (например, раз в неделю) вычисляется количество
    голосов $V$ и коэффициент $k$ для каждого деяния.

3.5. Блокчейн-инфраструктура
----------------------------

-   **Сеть:** Polygon (Proof-of-Stake) – низкие комиссии и
    экологичность.

-   **Хеширование:** раз в месяц вычисляются хеши (SHA-256) от Таблицы
    деяний и Таблицы голосов. Хеши публикуются в смарт-контракте с
    использованием мультиподписи Федерального совета (3 из 5).

-   **Токен KIT:** soulbound ERC-721 (невозвратный NFT) выпускается
    каждому участнику при вступлении. Метаданные токена хранятся в IPFS
    и содержат текущий ICD, индекс влияния, статус, поколение и ссылку
    на последний хеш таблицы. Обновление метаданных происходит по
    запросу (после подтверждения старостой).

-   **Защита:** мультиподпись, открытость данных, регулярные аудиты,
    слэшинг (блокировка токена) при доказанных фальсификациях.

3.6. Оценка влияния на мировой уровень
--------------------------------------

На основе накопленных данных ежемесячно вычисляются:

-   средний ICD движения;

-   глобальный УЦ;

-   динамика по поколениям.

Эти показатели публикуются на сайте движения и служат индикатором
прогресса.

Каждый участник уже является Хозяином по факту своего существования.
Задача движения — помочь увидеть свою повседневную жизнь, учёбу,

обыденное действие (написание курсовой, утренняя зарядка, помощь другу)
приближает человечество к величию.

Мы — наблюдатели, которые фиксируют не только внешние события, но и

1.3. Философия «достаточно просто быть» в деталях
-------------------------------------------------

1.3.1. Интерпретация как основной метод
---------------------------------------

Вместо директивных списков обязательных дел мы предлагаем каждому
участнику **интерпретировать** свои текущие занятия в терминах пяти
лучей (E, W, J, S, C). Например:

-   Студент-программист пишет диплом по компьютерному зрению. Он может
    интерпретировать это как вклад в луч **C** (культура/знания), а если
    его алгоритмы помогут сортировать мусор — то и в **S** (экология).

-   Спортсмен, устанавливающий личный рекорд в беге, не просто улучшает
    результат, а расширяет горизонты представлений о возможностях
    человеческого тела (**E**) и вдохновляет других (**C**).

-   Волонтёр, помогающий в приюте для животных, работает на
    лучи **W** (благо) и **S** (экология).

Интерпретация не требует дополнительных усилий — она меняет лишь угол
зрения, превращая рутину в осмысленный вклад в Величие.

результатами своей интерпретации.

курсовая по физике влияет на программу “Жизнь до 150 лет”» или «моё
спортивное достижение доказывает, что ресурс человека больше, чем
принято считать».  
Совет — это ритуал, укрепляющий общее смысловое поле и позволяющий

1.4. Проверочный вопрос: «Правда ли мы то самое человечество, которое может?»
-----------------------------------------------------------------------------

Этот вопрос — философский камертон движения. Каждое деяние, каждое
решение, каждую интерпретацию мы пропускаем через него. Он возвращает
нас к фундаментальной рефлексии:

-   Действительно ли мы, как вид, способны жить в гармонии с планетой?

-   Можем ли мы обеспечить каждого родившегося базовыми благами?

-   Готовы ли мы выйти к звёздам, сохранив душу?

Вопрос не имеет однозначного ответа, но сам процесс его постановки
превращает обыденность в экзистенциальный выбор. Он не даёт успокоиться
на достигнутом и постоянно напоминает: величие не даётся раз и навсегда,
его нужно подтверждать каждым днём.

1.5. Связь миссии с пятилучевой системой
----------------------------------------

Пять лучей (E, W, J, S, C) — это не просто набор активностей,
а **проекция философии Хозяина** на разные сферы жизни:

| Луч                    | Проекция философии                                                                               |
|------------------------|--------------------------------------------------------------------------------------------------|
| **E** (Энергия)        | Хозяин своего тела: забота о здоровье, силе, выносливости как фундаменте любой деятельности.     |
| **W** (Благо)          | Хозяин своего влияния: использование ресурсов для помощи другим, создание общего благосостояния. |
| **J** (Справедливость) | Хозяин своих прав и обязанностей: обеспечение равенства возможностей, защита достоинства.        |
| **S** (Экология)       | Хозяин планеты: ответственное отношение к природе, уборка, восстановление экосистем.             |
| **C** (Культура)       | Хозяин духа: творчество, познание, сохранение и приумножение культурного кода.                   |

Таким образом, быть Хозяином — значит гармонично развиваться во всех
этих направлениях, не оставляя «запущенных комнат» в своём внутреннем
доме.

1.6. Образ будущего: от УЦ-1 к УЦ-2 и дальше
--------------------------------------------

Мы видим свою миссию в том, чтобы ускорить переход человечества
к **УЦ-2** — уровню, где каждый родившийся гарантированно обеспечен
базовыми благами (питание, жильё, медицина, образование). Движение «Мы и
Наш Дом» готовит лидеров, которые понесут эти ценности в свои
профессиональные сферы, в свои семьи, в свои страны.  
Через сто лет мы хотим оглянуться и увидеть мир, где:

-   энергия чиста и доступна;

-   помощь ближнему — норма, а не подвиг;

-   справедливость закреплена в институтах;

-   экология стала естественной основой экономики;

-   культура и наука процветают.

И тогда мы сможем ответить на проверочный вопрос: **«Да, мы то самое
человечество, которое может. И мы это доказали»**.

**Резюме раздела**  
Миссия движения «Мы и Наш Дом» — формирование осознанного,
ответственного и гармоничного поколения, которое через интерпретацию
своей повседневной и научной деятельности строит новую эпоху духовно
развитого человечества. Философия «достаточно просто быть» снимает
барьер входа, делая участие естественным продолжением жизни, а Совет и
проверочный вопрос наполняют эту жизнь глубинным смыслом. Всё это вместе
служит одной цели — переходу на новый цивилизационный уровень, где
базовые блага гарантированы каждому, а человек становится подлинным
Хозяином своей судьбы и планеты.

2. Индекс Цивилизационного Развития (ICD): обоснование, расчет и визуализация
-----------------------------------------------------------------------------

2.1. Зачем нужен Индекс Цивилизационного Развития?
--------------------------------------------------

Индекс Цивилизационного Развития (ICD) — это интегральный показатель,
который позволяет:

-   **оценивать** гармоничность развития личности в пяти ключевых
    сферах, отражающих общечеловеческие ценности;

-   **мотивировать** участников движения к сбалансированному росту (не
    только в учёбе, но и в спорте, волонтёрстве, творчестве,
    экологической активности);

-   **сравнивать** вклад разных людей и поколений в общий прогресс
    человечества;

-   **отслеживать** динамику движения и приближение к цивилизационным
    рубежам (например, к УЦ-2, где каждый человек обеспечен базовыми
    благами).

Идея ICD родилась из синтеза нескольких подходов:

-   **Шкала Кардашёва** (энергетические уровни цивилизации) дала
    понимание, что развитие можно измерять количественно.

-   **Индекс человеческого развития (ИЧР) ООН** показал важность
    многомерных оценок (здоровье, образование, доход).

-   **Концепция устойчивого развития** добавила экологическое измерение.

-   **Философия «Мы и Наш Дом»** дополнила картину ценностями
    справедливости, культуры и личной ответственности.

Пять лучей (E, W, J, S, C) выбраны как минимальный, но достаточный
набор, охватывающий физическое, социальное, этическое, экологическое и
духовное измерения человеческой жизни.

2.2. Пять лучей развития
------------------------

| Луч   | Название       | Что учитывает                                    | Простой пример                                          |
|-------|----------------|--------------------------------------------------|---------------------------------------------------------|
| **E** | Энергия        | Спорт, здоровье, бодрость, энергоэффективность   | Участие в забеге, зарядка, занятия на турнике           |
| **W** | Благо          | Волонтёрство, помощь другим, благотворительность | Помощь отстающему, уборка аудитории                     |
| **J** | Справедливость | Равенство, защита прав, этичность                | Участие в диспуте, помощь первокурснику в адаптации     |
| **S** | Экология       | Чистота, уборка, забота о природе                | Субботник, посадка дерева, раздельный сбор              |
| **C** | Культура       | Творчество, знания, искусство, наука             | Написал стихи, выступил на концерте, опубликовал статью |

2.3. Двухкомпонентная модель оценки
-----------------------------------

Для разделения понятий «личная гармоничность» и «вклад в развитие
сообщества» вводится два показателя:

1.  **Индивидуальный индекс гармоничности (ICD)** – отражает, насколько
    сбалансировано развит участник по всем пяти лучам. Вычисляется на
    основе **исходных баллов** (без учёта множителей популярности).

2.  **Индекс влияния (Impact Index)** – количественная мера реального
    воздействия деяний участника на других людей и на прогресс в целом.
    Учитывает как исходные баллы, так и количество полученных голосов от
    членов движения.

2.4. Балльная система для деяний
--------------------------------

2.4.1. Базовые баллы
--------------------

Каждое деяние приносит баллы по одному главному лучу (2 балла) и до двух
дополнительным (по 1 баллу). Для прорывных изобретений (молекулярный
синтезатор, термоядерный реактор и т.п.) баллы могут быть в 1000 раз
выше, что отражает их исключительную значимость. Решение о применении
множителя принимает Федеральный совет на основе внешней экспертизы.

2.4.2. Базовый уровень при вступлении
-------------------------------------

Каждый новый участник при регистрации автоматически получает **по 1
баллу в каждый из пяти лучей**. Это гарантирует, что ни у кого не будет
нулевых значений, и личный ICD никогда не обратится в ноль, даже если
человек пока не совершал активных деяний. Базовые баллы также участвуют
в формировании максимумов, но их величина мала и не искажает общую
картину.

2.4.3. Учёт социального влияния (голосование)
---------------------------------------------

Любой участник движения может проголосовать за деяние другого участника,
если считает, что оно оказало на него положительное влияние (вдохновило,
помогло в учёбе, изменило взгляды). Голосование проводится через
специальный сервис (Telegram-бот или веб-форму) и фиксируется в реестре.
(Например добавляется в VK и идёт голосование, список деяний
автоматически ранжируется и выводится общий рейтинг и другие факторы)

Для каждого деяния вычисляется **коэффициент социального влияния**:

$$k = 1 + \ln(1 + V),
$$

где $V$ – количество уникальных голосов, отданных за это деяние.
Логарифмическая шкала предотвращает неограниченный рост и даёт значимый
прирост при массовой поддержке (например, при $V = 1000$, $k \approx 8$;
при $V = 10^{6}$, $k \approx 14,5$).

**Взвешенные баллы** деяния по лучам, используемые только для расчёта
индекса влияния участника, получаются умножением исходных баллов на $k$:

$$\text{Взвешенные\ баллы} = \text{Исходные\ баллы} \times k.
$$

2.5. Расчёт индивидуального индекса гармоничности (ICD)
-------------------------------------------------------

2.5.1. Накопление исходных баллов
---------------------------------

Пусть участник имеет суммарные исходные баллы по каждому
лучу $B_{E},B_{W},B_{J},B_{S},B_{C}$ (включая базовый 1 балл за
вступление). Эти баллы не учитывают голосование.

2.5.2. Максимумы в движении
---------------------------

Обозначим $M_{E},M_{W},M_{J},M_{S},M_{C}$ – максимальные значения
исходных баллов среди **всех участников движения** на текущий момент.
Они автоматически обновляются при каждом новом деянии.

2.5.3. Нормированные значения
-----------------------------

Для участника нормированные значения по каждому лучу:

$$x_{l} = \frac{B_{l}}{M_{l}}(\text{при\ }M_{l} > 0).
$$

Так как все $B_{l} \geq 1$ и $M_{l} \geq 1$, значения лежат в
интервале $(0,1\rbrack$.

2.5.4. Индивидуальный ICD
-------------------------

Индивидуальный индекс гармоничности вычисляется как **среднее
геометрическое** нормированных значений:

$$\text{IC}D_{\text{личн}} = \sqrt[5]{x_{E} \cdot x_{W} \cdot x_{J} \cdot x_{S} \cdot x_{C}}.
$$

Среднее геометрическое резко падает, если хотя бы одно из значений мало,
что отражает требование гармоничного развития. Однако благодаря базовым
баллам ни один сомножитель не равен нулю, поэтому ICD всегда
положителен. Чем ближе к 1, тем более сбалансирован и развит участник.

2.6. Расчёт индекса влияния (Impact Index)
------------------------------------------

Индекс влияния участника учитывает все его деяния с весами, усиленными
голосованием:

$$I_{\text{влияния}} = \sum_{\text{деяния}}^{}\left( E_{\text{взв}} + W_{\text{взв}} + J_{\text{взв}} + S_{\text{взв}} + C_{\text{взв}} \right),
$$

где $E_{\text{взв}}$ – взвешенные баллы деяния по лучу E и т.д.

Этот индекс может достигать огромных значений для авторов прорывных
работ, получивших широкое признание. Он используется для присвоения
статусов:

-   **Мастер** – при превышении порога, например, 1000.

-   **Творец** – при превышении 5000.

-   **Легенда** – при превышении 10000.

Статусы присваиваются автоматически, без дополнительного голосования
Совета, как только сумма взвешенных баллов достигает порога.

2.7. Глобальный Уровень Цивилизации (УЦ)
----------------------------------------

Уровень цивилизации в целом определяется не по средним показателям, а
по **максимальным достижениям** человечества в каждом луче, что
соответствует логике шкалы Кардашёва.

2.7.1. Абсолютные реперы
------------------------

Для каждого луча вводятся фиксированные абсолютные
реперы $L_{l}^{\text{abs}}$ – значения, соответствующие гипотетическому
уровню УЦ-5 (полное развитие). Например:

-   $L_{E}^{\text{abs}}$ может соответствовать энергопотреблению,
    эквивалентному полному использованию солнечной энергии
    (около $3.8 \cdot 10^{26}$ Вт), но для практических целей можно
    выбрать более достижимые ориентиры (например, 1 ТВт на душу
    населения). Важно, чтобы реперы оставались неизменными для
    сохранения сравнимости во времени.

-   $L_{W}^{\text{abs}}$ – обеспечение всех базовых благ для 100%
    населения планеты.

-   $L_{J}^{\text{abs}}$ – полное равенство возможностей (индекс Джини =
    0).

-   $L_{S}^{\text{abs}}$ – нулевой углеродный след, полное
    восстановление экосистем.

-   $L_{C}^{\text{abs}}$ – доступ ко всем знаниям человечества,
    максимальная реализация творческого потенциала.

2.7.2. Нормированные максимумы
------------------------------

Текущие максимумы исходных баллов $M_{l}$ нормируются относительно
абсолютных реперов:

$$y_{l}(t) = \frac{M_{l}(t)}{L_{l}^{\text{abs}}}.
$$

2.7.3. Глобальный УЦ
--------------------

Глобальный уровень цивилизации вычисляется как среднее геометрическое
нормированных максимумов:

$$\text{УЦ}(t) = \sqrt[5]{y_{E}(t) \cdot y_{W}(t) \cdot y_{J}(t) \cdot y_{S}(t) \cdot y_{C}(t)}.
$$

Это значение лежит в интервале $\left\lbrack 0,1 \right\rbrack$ и
автоматически обновляется при любом улучшении максимума в одном из
лучей. Таким образом, появление молекулярного синтезатора или
термоядерного реактора сразу повышает соответствующий $M_{l}$ и,
следовательно, глобальный УЦ.

2.8. Связь с уровнями цивилизации
---------------------------------

Диапазоны глобального УЦ соотносятся с качественными уровнями:

| Уровень  | Название         | Диапазон УЦ | Характеристика                                                    |
|----------|------------------|-------------|-------------------------------------------------------------------|
| **УЦ-0** | Зарождающаяся    | 0,00 – 0,20 | Выживание, локальные группы, базовые блага не гарантированы       |
| **УЦ-1** | Планетарная      | 0,20 – 0,40 | Формирование глобального общества, базовые блага доступны не всем |
| **УЦ-2** | **Обеспеченная** | 0,40 – 0,60 | **Каждый родившийся гарантированно обеспечен базовыми благами**   |
| **УЦ-3** | Звёздная         | 0,60 – 0,80 | Управление энергией звезды, активное освоение космоса             |
| **УЦ-4** | Галактическая    | 0,80 – 0,90 | Контроль над энергией галактики, гармония с природой              |
| **УЦ-5** | Постчеловеческая | \>0,90      | Духовная зрелость, управление пространством-временем              |

2.9.2. Общий рейтинг (таблица лидеров)

-----------------------------

График изменения среднего ICD для разных поколений (по году вступления)

2.9.4. Доска почёта

В каждом вузе оформляется стенд с фотографиями лучших участников (по

2.9.5. Инфографика для социальных сетей
---------------------------------------

Ежемесячно публикуются «звёзды месяца» – участники с наибольшим
приростом индекса влияния или достигшие новых статусов.

2.10. Пример расчёта с учётом новых вводных
-------------------------------------------

Рассмотрим гипотетического участника Анну, создавшую молекулярный
синтезатор. При вступлении она получила по 1 баллу в каждый луч. Затем
её деяние принесло исходные баллы: W=1000, S=500, C=500. Также за деяние
проголосовало 8000 человек.
Коэффициент $k = 1 + \ln(8001) \approx 9,987$. Взвешенные баллы: W=9987,
S=4993.5, C=4993.5. Предположим, что до её деяния максимумы были:
M\_E=400, M\_W=350, M\_J=200, M\_S=300, M\_C=450. После деяния новые
максимумы: M\_W=1000, M\_S=500, M\_C=500, остальные не изменились.

Нормированные значения Анны:

-   $x_{E} = (1)/400 = 0.0025$

-   $x_{W} = (1000 + 1)/1000 = 1.001/1000?$ Надо уточнить: B\_W =
    исходные баллы Анны по W = 1000 (деяние) + 1 (базовый) = 1001. Но
    M\_W стало 1000? Но максимум должен быть не меньше, чем у Анны,
    поэтому M\_W должно быть 1001. Ошибка: при добавлении деяния с
    баллами 1000, новый максимум становится 1001 (если базовый
    учитывается). Но базовый у всех есть, значит максимум может быть
    больше 1000. Логично: максимум считается по полным исходным баллам
    участников, включая базовые. Тогда после деяния Анны её B\_W=1001, и
    M\_W становится 1001. Аналогично B\_S=501, M\_S=501; B\_C=501,
    M\_C=501. Тогда нормированные:

-   x\_E = 1/400 = 0.0025

-   x\_W = 1001/1001 = 1

-   x\_J = 1/200 = 0.005

-   x\_S = 501/501 = 1

-   x\_C = 501/501 = 1  
    Произведение = 0.0025 \* 1 \* 0.005 \* 1 \* 1 = 1.25e-5, корень
    пятой степени ≈ (1.25e-5)^0.2 = 10^{-4} \* (1.25)^0.2 ≈ 0.0001 \*
    1.045 = 0.0001045. Это очень мало из-за низких значений E и J. ICD
    Анны ≈ 0.0001, что отражает её негармоничность (отсутствие спорта и
    справедливости). Индекс влияния = 9987+4993.5+4993.5 ≈ 19974, что
    автоматически даёт статус Легенды. Глобальный УЦ обновится за счёт
    новых максимумов: y\_E=400/L\_E, y\_W=1001/L\_W и т.д.

Таким образом, система работает: уникальный вклад признаётся через
влияние, а личная гармоничность остаётся низкой, стимулируя Анну
развиваться в других направлениях.

**Резюме раздела**  
Индекс Цивилизационного Развития (ICD) в его двухкомпонентной форме
позволяет объективно оценивать как гармоничность личности, так и её
реальное влияние на прогресс. Базовые баллы гарантируют отсутствие
нулевых значений, а учёт голосования через логарифмический коэффициент
даёт автоматическое признание прорывных достижений. Глобальный УЦ,
привязанный к абсолютным реперам, автоматически отражает рост максимумов
и позволяет отслеживать движение человечества к высшим уровням
цивилизации. Визуализация делает эти показатели наглядными и
мотивирующими для всех участников.

3. Учёт деяний и цифровой след
------------------------------

Система учёта деяний – это «кровеносная система» движения «Мы и Наш
Дом». Она обеспечивает прозрачность, неизменность и мотивацию, фиксируя
вклад каждого участника в пяти лучах развития, а также учитывая
социальное влияние его деяний через механизм голосования. Главные
принципы: **неизменность** (никакие записи нельзя подделать или задним
числом исправить), **прозрачность** (любой желающий может проверить
данные), **простота** (студенты не должны быть экспертами в блокчейне)
и **справедливость учёта влияния** (голоса защищены от накруток).

3.1. Принципы учёта
-------------------

-   **Добровольность** – участник сам (или с помощью старосты)
    инициирует запись своего деяния.

-   **Подтверждённость** – каждое деяние должно иметь простое
    доказательство (фото, ссылка, свидетельство второго участника).

-   **Базовый уровень** – каждый новый участник при регистрации
    автоматически получает по 1 баллу в каждый из пяти лучей, что
    гарантирует ненулевые значения и служит основой для индекса
    гармоничности.

-   **Учёт социального влияния** – любой участник может проголосовать за
    деяние другого, если оно оказало на него положительное воздействие.
    Голоса фиксируются в отдельном реестре и влияют на индекс влияния.

-   **Неизменность истории** – все записи, однажды внесённые в реестры,
    не могут быть изменены; любые исправления возможны только через
    добавление новой записи с комментарием.

-   **Открытость** – итоговые данные (анонимизированные при
    необходимости) доступны всем членам движения.

3.2. Алгоритм регистрации деяния
--------------------------------

Процесс построен так, чтобы минимизировать бюрократию и использовать
привычные студентам инструменты (Google Таблицы, Telegram).

3.2.1. Инициация
----------------

Участник (или староста группы) заполняет **короткую форму** в Google
Forms или в Telegram-боте, указывая:

-   ФИО (или ID в реестре);

-   дату деяния;

-   краткое описание (например, «участие в субботнике в парке
    Горького»);

-   главный луч (E, W, J, S, C) и, при необходимости, дополнительные
    лучи (до двух);

-   ссылку на подтверждение (фото, пост в соцсетях, скриншот переписки и
    т.п.).

Для прорывных деяний (изобретений, открытий) предусмотрена особая
процедура: заявка подаётся в Федеральный совет с приложением экспертных
заключений, патентов, публикаций. Совет назначает внешнюю экспертизу и
утверждает множитель исходных баллов (например, ×1000).

Пояснение к пункту 3.2.1: необходимость особой процедуры для прорывных деяний
-----------------------------------------------------------------------------

Вопрос закономерен: действительно, можно было бы положиться на механизм
голосования (индекс влияния) как на естественный измеритель значимости
деяния. Однако в нашей системе мы разделяем два принципиально разных
аспекта: **объективную ценность деяния для прогресса
человечества** и **его социальное признание**. Для этого и введены две
независимые метрики – исходные баллы (с возможным экспертным множителем)
и взвешенные баллы (с учётом голосов). Рассмотрим, почему для прорывных
деяний требуется особая процедура утверждения.

**1. Масштаб вклада требует немедленного отражения в глобальных
максимумах**

Прорывные изобретения (термоядерный реактор, молекулярный синтезатор)
имеют потенциал изменить ход цивилизации. Их влияние на глобальные
максимумы $M_{l}$ и, следовательно, на **Уровень Цивилизации
(УЦ)** должно быть отражено **немедленно**, ещё до того, как они получат
массовое признание в сообществе. Если ждать, пока тысячи или миллионы
людей проголосуют, это может занять годы, а нам нужно актуальное
состояние максимумов для расчёта УЦ. Например, создание работающего
термоядерного реактора сразу поднимает энергетическую планку
человечества, и это должно быть зафиксировано в индексе независимо от
того, сколько студентов узнали об этом событии.

**2. Объективность versus популярность**

Голосование отражает субъективное мнение участников, которое может
зависеть от моды, пиара, эмоций или даже целенаправленных манипуляций.
Прорывное деяние может быть:

-   **непонято современниками** (как многие великие открытия в истории)
    и получить мало голосов;

-   **переоценено** из-за временного хайпа, не соответствующего реальной
    научной ценности.

Экспертная оценка (с привлечением внешних специалистов, патентов,
рецензируемых публикаций) даёт более объективную базу для определения
исходного веса деяния. Она опирается на признанные научные и
профессиональные критерии, а не на текущие настроения сообщества.

**3. Защита от злоупотреблений и инфляции**

Без экспертизы любой участник мог бы заявить о «прорыве» и потребовать
множитель ×1000, искажая всю систему. Даже при наличии голосования,
накрутка голосов (например, через ботов или «круговую поруку») может
создать ложное впечатление значимости. Экспертиза служит **фильтром**,
который подтверждает реальную научную или технологическую ценность и
предотвращает размывание показателей.

**4. Синергия двух механизмов**

Экспертный множитель и голосование работают **вместе**, а не вместо друг
друга:

-   **Исходные баллы (с множителем)** фиксируют объективный вклад деяния
    в прогресс человечества. Они сразу влияют на максимумы $M_{l}$ и,
    через них, на глобальный УЦ.

-   **Взвешенные баллы (с учётом голосов)** отражают социальное влияние
    деяния – сколько людей оно вдохновило, изменило их взгляды, помогло
    в развитии. Это влияет на индекс влияния участника и его статус.

Таким образом, даже если деяние получило высокий экспертный множитель,
но осталось неизвестным (мало голосов), его автор не получит огромный
индекс влияния, что справедливо. И наоборот, популярное, но не прорывное
деяние (например, вирусное видео) может набрать много голосов, но его
исходные баллы останутся скромными, и оно не поднимет глобальные
максимумы.

**5. Пример для иллюстрации**

Представьте, что изобретение термоядерного реактора произошло в 2040
году. Без экспертной оценки мы бы ждали, пока сообщество осознает его
значение и проголосует. Но даже если проголосуют 100 000 человек, их
голоса дадут коэффициент $k \approx 1 + \ln(100001) \approx 12,5$. Тогда
взвешенные баллы автора станут огромными, но максимумы исходных
баллов $M_{E}$ останутся на прежнем уровне (например, 400 от предыдущих
спортивных достижений). Глобальный УЦ не изменится, что абсурдно, ведь
технологический скачок уже произошёл.

При экспертной процедуре мы сразу устанавливаем исходные баллы,
скажем, $E = 1000$ (с множителем). Тогда $M_{E}$ становится 1000, и УЦ
мгновенно отражает новый энергетический уровень. А голосование позже
добавит автору индекс влияния, укрепив его статус.

**Вывод**

Особая процедура для прорывных деяний необходима, чтобы:

-   своевременно и объективно учитывать их вклад в глобальное развитие
    (максимумы и УЦ);

-   отделять реальные достижения от временной популярности;

-   защищать систему от манипуляций и инфляции показателей.

Голосование же выполняет свою важную роль – измеряет социальное влияние,
но не может заменить экспертную оценку масштаба деяния. Оба механизма
дополняют друг друга, обеспечивая справедливость и точность интегральной
оценки.

3.2.2. Валидация
----------------

-   **Староста группы** (назначенный из числа Мастеров) проверяет заявку
    в течение 3–5 дней. Он может запросить дополнительные
    доказательства.

-   Если заявка корректна, староста утверждает её в той же форме или в
    общей таблице.

-   В спорных случаях (например, неясно, какой луч главный) вопрос
    выносится на **Совет первички**.

3.2.3. Внесение в реестр исходных баллов
----------------------------------------

Все утверждённые заявки попадают в **единую Google Таблицу
«Деяния»** (далее – Таблица деяний) с разграничением доступа: только
старосты и кураторы имеют право редактирования, остальные – только
чтение. Таблица содержит столбцы:

-   ID участника

-   ФИО

-   Дата деяния

-   Описание

-   Исходные баллы по лучам (E, W, J, S, C) – согласно правилам (базовые
    2/1, умноженные на множитель для прорывов)

-   Ссылка на подтверждение

-   Кем подтверждено (староста)

-   Метка времени внесения записи

Пример записи:

| ID  | ФИО         | Дата       | Описание       | E   | W   | J   | S   | C   | Подтверждение | Староста   |
|-----|-------------|------------|----------------|-----|-----|-----|-----|-----|---------------|------------|
| 001 | Иванов Иван | 05.03.2026 | Плоггинг-забег | 1   | 0   | 0   | 2   | 0   | \[ссылка\]    | Петрова А. |

3.2.4. Базовые баллы при вступлении
-----------------------------------

При регистрации нового участника в Таблицу деяний автоматически вносится
запись о начислении 1 балла в каждый луч (можно с датой вступления и
пометкой «базовый»). Это гарантирует, что даже у новичков все $B_{l}$ ≥
1.

3.3. Сервис голосования и учёт влияния
--------------------------------------

3.3.1. Голосование за деяние
----------------------------

Каждый участник (обладатель действующего токена KIT) может проголосовать
за любое деяние другого участника, если считает, что оно оказало на него
положительное влияние (вдохновило, помогло, изменило взгляды).
Голосование осуществляется через специальный интерфейс (Telegram-бот или
веб-форму), где указывается ID деяния. Одно деяние можно оценить только
один раз; повторные голоса от того же участника не засчитываются.

3.3.2. Реестр голосов
---------------------

Все голоса записываются в отдельную **Google Таблицу «Голоса»** со
столбцами:

-   ID деяния

-   ID голосующего

-   Дата и время голоса

-   Хеш транзакции (если используется блокчейн для дополнительной
    защиты)

3.3.3. Расчёт коэффициента социального влияния
----------------------------------------------

Периодически (например, раз в неделю) для каждого деяния вычисляется
количество уникальных голосов $V$ и коэффициент:

$$k = 1 + \ln(1 + V).
$$

Взвешенные баллы деяния по каждому лучу получаются умножением исходных
баллов (из Таблицы деяний) на $k$:

$$\text{Взвешенные\ баллы}_{l} = \text{Исходные\ баллы}_{l} \times k.
$$

Эти взвешенные баллы используются только для расчёта **индекса
влияния** участника и нигде больше.

3.3.4. Хранение взвешенных баллов
---------------------------------

Взвешенные баллы могут не храниться постоянно, а вычисляться на лету при
запросе. Для оптимизации можно создать таблицу с агрегированными
значениями, обновляемую после каждого пересчёта.

3.4. Защита от накруток и дискредитации
---------------------------------------

-   **Идентификация голосующего**: голосовать может только участник с
    валидным токеном KIT, что исключает ботов и множественные аккаунты.

-   **Мониторинг аномалий**: автоматические скрипты выявляют
    подозрительные паттерны (например, массовые голоса от новых
    участников за одно деяние, голоса «круговой поруки»).

-   **Аудит Советом**: при обнаружении аномалий проводится проверка.
    Если подтверждается накрутка, голоса аннулируются, а нарушители
    могут быть временно заблокированы или подвергнуты слэшингу (см. п.
    3.6.3).

-   **Прозрачность**: любой участник может запросить выгрузку голосов и
    провести собственный анализ.

3.5. Блокчейн-инфраструктура
----------------------------

3.5.1. Запись хешей
-------------------

Для обеспечения неизменности данных периодически (например, раз в месяц)
вычисляются хеши (SHA-256) от:

-   Таблицы деяний (все записи, включая базовые);

-   Таблицы голосов.  
    Полученные хеши публикуются в смарт-контракте сети Polygon (или
    другой PoS-сети) с использованием мультиподписи Совета (требуется 3
    из 5 подписей). Адрес контракта и интерфейс для проверки публикуются
    на сайте движения.

3.5.2. Хранение таблиц
----------------------

Сами таблицы хранятся в Google Drive с открытым доступом для членов
движения. Для дополнительной децентрализации можно дублировать их в
IPFS, а CID также включать в транзакцию.

3.5.3. Слэшинг (наказание)
--------------------------

В случае обнаружения фальсификации (например, староста подтвердил
несуществующее деяние) Совет может принять решение о «сжигании» токена
нарушителя (блокировка, перевод в специальный адрес). Информация о
нарушении заносится в публичный реестр. Для голосований с накруткой
применяется аннулирование голосов и временная блокировка права голоса.

3.6. Цифровой токен KIT (Keeper Impact Token)
---------------------------------------------

3.6.1. Назначение и стандарт
----------------------------

KIT – soulbound-токен (невозвратный NFT) в сети Polygon, выпускаемый
каждому участнику при вступлении. Он служит:

-   идентификатором в цифровой среде;

-   хранилищем агрегированной информации о прогрессе (текущий ICD,
    индекс влияния, статус);

-   средством верификации достижений при переходе в другой вуз или
    трудоустройстве.

**3.6.2. Структура метаданных**

Метаданные токена хранятся в IPFS и обновляются при изменении ключевых
показателей (раз в семестр или по запросу). Они содержат:

-   имя участника (или псевдоним);

-   год вступления (поколение);

-   текущий ICD (рассчитывается по формуле из п. 2.5.4);

-   индекс влияния (сумма взвешенных баллов);

-   статус (Хранитель, Мастер, Творец, Легенда), определяемый порогами
    индекса влияния;

-   ссылку на хеш последней версии таблицы деяний (для детальной
    проверки);

-   хеш предыдущей версии метаданных (для обеспечения цепочки).

Пример метаданных (JSON):

json

{

"name": "Ivan Ivanov",

"generation": 2026,

"icd": 0.73,

"impactIndex": 15420,

"status": "Легенда",

"historyRef": "bafybeia...",

"previousVersion": "bafkreih..."

}

3.6.3. Процесс минтинга и обновления
------------------------------------

-   **Минтинг**: при регистрации нового участника Совет утверждает его
    заявку, и бот (или вручную) вызывает функцию mint с указанием адреса
    кошелька участника (кошелёк можно создать через простой мобильный
    кошелёк типа MetaMask или даже через бота с использованием
    смарт-контракта без газа).

-   **Обновление метаданных**: раз в семестр или при достижении нового
    статуса участник может запросить обновление. Староста или куратор
    подтверждают новые значения ICD и индекса влияния, после чего по
    мультиподписи вызывается функция updateMetadata, изменяющая URI
    токена.

3.7. Оценка влияния на мировой уровень развития человечества
------------------------------------------------------------

Накопленные данные позволяют агрегировать индивидуальные показатели и
отслеживать прогресс сообщества и человечества в целом.

3.7.1. Глобальный ICD движения
------------------------------

Вычисляется как среднее арифметическое индивидуальных ICD всех
участников. Публикуется ежемесячно.

3.7.2. Глобальный Уровень Цивилизации (УЦ)
------------------------------------------

Рассчитывается по формуле из п. 2.7.3 на основе текущих максимумов
исходных баллов $M_{l}$ и абсолютных реперов. Значение УЦ автоматически
обновляется при любом улучшении максимума (например, после регистрации
прорывного деяния).

3.7.3. Визуализация прогресса
-----------------------------

На сайте движения отображаются:

-   график роста глобального ICD и УЦ во времени;

-   «карта пяти лучей» с текущими максимумами;

-   прогноз достижения следующего уровня УЦ при сохранении темпов.

3.8. Динамика поколений
-----------------------

Система позволяет анализировать вклад разных возрастных когорт.

-   Каждому участнику присваивается номер поколения – год его первого
    деяния (или год вступления).

-   Раз в год вычисляются средний ICD и средний индекс влияния для
    каждой генерации.

-   Строятся графики, показывающие, как меняется активность и
    гармоничность от поколения к поколению.

-   Эти данные помогают корректировать наставническую работу и
    мотивационные программы.

Пример графика (текстовое описание):

Поколение 2026 года через 3 года достигло среднего ICD 0.48 и индекса
влияния 1200, а поколение 2027 года уже через год показывает ICD 0.42 и
влияние 800 – вероятно, за счёт улучшенной системы наставничества.

**Резюме раздела**  
Учёт деяний построен на сочетании простых цифровых инструментов (Google
Таблицы) и криптографической защиты (блокчейн, soulbound-токены).
Введение двух реестров (деяний и голосов) позволяет объективно
фиксировать как сам вклад, так и его социальное влияние. Базовые баллы
гарантируют ненулевые значения, а механизм голосования с логарифмическим
коэффициентом автоматически отражает резонанс прорывных достижений.
Агрегированные показатели (глобальный ICD, УЦ, динамика поколений) не
только мотивируют участников, но и позволяют движению отслеживать свой
вклад в восхождение человечества к новым цивилизационным уровням. Токен
KIT выполняет функцию «цифровой души», сохраняя историю достижений и
статус человека на всю жизнь.

4. Для чего нужно измерять социальное влияние и что это даёт автору деяния?
---------------------------------------------------------------------------

В системе оценки движения «Мы и Наш Дом» мы сознательно разделили два
разных аспекта любого деяния:

-   **объективную ценность** (научную, технологическую, социальную),
    которая измеряется исходными баллами и при необходимости усиливается
    экспертным множителем;

-   **социальное влияние** – то, как деяние воспринято другими людьми,
    скольких оно вдохновило, помогло, изменило их мышление или
    поведение.

Индекс влияния (Impact Index) как раз и отражает это второе измерение.
Давайте разберёмся, зачем оно нужно и что даёт автору.

4.1. Что измеряет индекс влияния?
---------------------------------

Индекс влияния вычисляется на основе **голосов участников движения** за
конкретное деяние. Каждый голос – это сигнал: «Это деяние повлияло на
меня лично, оно важно для меня».
Коэффициент $k = 1 + \ln(1 + V)$ превращает количество голосов в
умеренный множитель, не позволяя популярности полностью затмить
объективную ценность, но давая значимый прирост при массовой поддержке.

Таким образом, индекс влияния – это **мера резонанса**, которую деяние
вызывает в сообществе.

4.2. Что индекс влияния даёт автору?
------------------------------------

4.2.1. Признание и статус в сообществе
--------------------------------------

В движении введены статусы (Мастер, Творец, Легенда), которые
присваиваются **автоматически** при достижении определённых порогов
индекса влияния. Это значит, что автор, чьи деяния получили широкий
отклик, становится признанным лидером мнений, наставником, примером для
подражания. Ему не нужно дополнительно «пробивать» своё признание через
Совет – система сама фиксирует его вклад в умы и души людей.

4.2.2. Возможность влиять на других и вести за собой
----------------------------------------------------

Высокий индекс влияния открывает доступ к **роли наставника**: такие
участники могут официально вести группы новичков, делиться опытом,
помогать другим развиваться. Они становятся теми, кто формирует культуру
движения, передаёт ценности следующим поколениям. Это не только почётно,
но и даёт реальный опыт управления и педагогики, который ценится в любой
карьере.

4.2.3. Участие в управлении движением
-------------------------------------

В перспективе, когда движение вырастет, индекс влияния может стать
основой для **взвешенного голосования** в Палате вклада (если такая
будет создана). Те, кто доказал свою способность влиять на людей,
получают больший вес при принятии стратегических решений – это
справедливо и повышает качество управления.

4.2.4. Нематериальные бонусы и привилегии
-----------------------------------------

Авторы с высоким индексом влияния получают:

-   Право **предлагать собственные проекты** и получать для них ресурсы
    (например, бронирование аудиторий, помощь в организации
    мероприятий).

-   Место на **Доске почёта** (физической и цифровой), что повышает их
    авторитет в вузе и может быть замечено преподавателями и
    администрацией.

-   Возможность **публиковать свои материалы** в официальных каналах
    движения, становясь recognised экспертом в своей области.

-   Рекомендательные письма от движения, подкреплённые конкретными
    цифрами (индекс влияния, количество вдохновлённых людей) – это
    мощный бонус при поступлении в магистратуру или при трудоустройстве.

4.2.5. Обратная связь и мотивация
---------------------------------

Знание того, что твои деяния реально кого-то затронули, вдохновили,
помогли – само по себе мощный стимул. Индекс влияния даёт автору
количественное подтверждение его значимости для других. Это особенно
важно для тех, кто занимается просветительской, творческой, научной
деятельностью: видеть, что твой труд не остаётся незамеченным, что он
меняет мир вокруг.

4.3. Зачем это движение?
------------------------

4.3.1. Выявление лидеров и наставников
--------------------------------------

Индекс влияния помогает естественным образом выявить тех, кто умеет не
только делать, но и увлекать, вдохновлять, передавать знания. Эти люди
становятся опорой движения, его «критической массой», обеспечивающей
устойчивость и развитие.

4.3.2. Стимулирование популяризации и обмена знаниями
-----------------------------------------------------

Зная, что голоса влияют на статус, авторы будут больше рассказывать о
своих деяниях, делиться опытом, проводить открытые лекции,
мастер-классы. Это создаёт здоровую среду обмена знаниями и ускоряет
распространение лучших практик.

4.3.3. Укрепление сообщества
----------------------------

Голосование за деяния – это акт благодарности и признания. Оно создаёт
связи между участниками, формирует культуру взаимной поддержки. Когда
человек видит, что его труд оценён, он чувствует себя частью чего-то
большего, что повышает лояльность и удерживает в движении.

4.3.4. Объективная обратная связь для автора
--------------------------------------------

Индекс влияния – это индикатор того, насколько деятельность автора
соответствует потребностям и интересам сообщества. Если деяние набирает
мало голосов, автор может задуматься: возможно, стоит изменить форму
подачи, больше общаться с людьми, объяснять ценность своей работы.

4.4. Почему нельзя заменить индекс влияния экспертным множителем?
-----------------------------------------------------------------

Экспертная оценка (например, при прорывных изобретениях)
фиксирует **объективную значимость** деяния для прогресса человечества.
Но она не измеряет его **социальный резонанс**. Может быть великое
открытие, о котором знают лишь несколько специалистов, и его влияние на
умы масс пока ничтожно. И наоборот, популярный научно-популярный пост
может вдохновить тысячи людей, хотя его объективная научная ценность
невелика. Оба аспекта важны: одно без другого даёт неполную картину.

-   **Высокий экспертный множитель + низкий индекс влияния** – автор
    внёс огромный вклад в прогресс, но пока не сумел донести его до
    сообщества. Его статус Легенда он получит за объективные заслуги, но
    как наставник и лидер мнений может быть не востребован, пока не
    начнёт активно общаться с людьми.

-   **Низкий экспертный множитель + высокий индекс влияния** – автор,
    возможно, не совершил прорыва, но он отличный популяризатор,
    учитель, вдохновитель. Он может стать Мастером или Творцом и играть
    ключевую роль в воспитании новых поколений Мы и Наш Дом.

4.5. Пример из жизни
--------------------

Представим двух участников:

-   **Анна** – создала термоядерный реактор. Её деяние получает
    экспертный множитель ×1000, но она не очень общительна, редко
    выступает, не ведёт соцсети. Её индекс влияния невелик (скажем, 200
    голосов). Она получит огромные исходные баллы, максимумы вырастут,
    но её статус в движении будет просто «Легенда» (по порогу индекса
    влияния она может не дотянуть до Легенды, если порог высок). Однако
    она всё равно будет признана сообществом, но её влияние на умы пока
    ограничено.

-   **Пётр** – ведёт популярный блог о науке, его видео о термоядерном
    синтезе набирают тысячи голосов. Сам он не изобретатель, но его
    индекс влияния огромен. Он становится Мастером (или даже Творцом),
    его приглашают наставником, он участвует в управлении, его мнение
    уважают. Он реально влияет на то, как многие воспринимают науку и
    ценности движения.

Оба важны, и система справедливо оценивает их по-разному.

**Итог**

Индекс влияния даёт автору:

-   **статус и признание** в сообществе;

-   **возможность влиять** на других и вести за собой;

-   **нематериальные привилегии** (доска почёта, доступ к ресурсам,
    рекомендации);

-   **обратную связь** о том, насколько его деятельность находит отклик.

Для движения это инструмент выявления лидеров, стимулирования обмена
знаниями и укрепления внутренних связей. Вместе с объективной оценкой
(исходные баллы и экспертный множитель) индекс влияния создаёт полную
картину вклада человека в развитие человечества – и как творца нового, и
как вдохновителя других.

5. Организационная структура движения
-------------------------------------

5.1. Принципы построения
------------------------

Организационная структура движения «Мы и Наш Дом» разработана с учётом
необходимости:

-   **масштабируемости** – от одной студенческой группы до всероссийской
    (и в перспективе международной) сети;

-   **прозрачности и демократичности** – все ключевые решения
    принимаются выборными органами с участием активных членов;

-   **эффективности** – чёткое разделение полномочий, минимизация
    бюрократии, использование цифровых инструментов;

-   **стимулирования активности** – статусная система напрямую связана с
    возможностью занимать руководящие должности.

Структура строится по **матричному принципу**: вертикаль управления
(федеральный центр – регионы – вузы) дополняется горизонтальными связями
(кураторы пяти лучей, проектные группы, наставничество). Это
обеспечивает как единство стратегии, так и гибкость в реализации
локальных инициатив.

5.2. Уровни организационной структуры
-------------------------------------

5.2.1. Первичная организация (вузовский уровень)
------------------------------------------------

Первичная организация создаётся в каждом вузе, где есть не менее 5
участников. Она является основной ячейкой движения, где происходит
повседневная деятельность, учёт деяний, проведение Советов и
наставничество.

**Состав первичной организации:**

-   **Все участники** данного вуза (обладатели токенов KIT). Они
    образуют **Общее собрание** – высший орган первички, собирающийся не
    реже одного раза в семестр для выборов Совета и утверждения отчётов.

-   **Совет первичной организации** (от 3 до 7 человек) – исполнительный
    орган, избираемый Общим собранием сроком на один год. В состав
    Совета входят:

    -   **Лидер первички** – координирует работу, представляет
        организацию в региональном совете, отвечает за выполнение
        решений вышестоящих органов.

    -   **Старосты курсов** (по одному от каждого курса, где есть
        участники) – отвечают за сбор заявок на деяния, первичную
        валидацию, связь с кураторами лучей.

    -   **Кураторы лучей** (E, W, J, S, C) – назначаются Советом из
        числа Мастеров и выше; развивают соответствующее направление в
        вузе, помогают участникам интерпретировать деяния, организуют
        тематические мероприятия.

    -   **Секретарь** – ведёт документацию, отвечает за взаимодействие с
        цифровыми инструментами (таблицы, бот).

**Функции первичной организации:**

-   Регистрация новых участников и помощь в создании кошельков.

-   Приём и первичная валидация заявок на деяния (через старост).

-   Проведение ежемесячных **Советов первички**, где участники делятся
    интерпретациями своих деяний, обсуждают общие проекты, выдвигают
    инициативы.

-   Организация внутривузовских мероприятий (спортивные соревнования,
    субботники, лекции, концерты) с целью вовлечения новых студентов.

-   Назначение наставников для новичков из числа Мастеров.

-   Ведение внутривузовской Доски почёта.

-   Подготовка отчётов для регионального совета.

**Взаимодействие с регионом:** Лидер первички входит в состав
регионального совета (с правом голоса) и участвует в ежемесячных
координационных встречах.

5.2.2. Региональный уровень
---------------------------

Региональные советы создаются в федеральных округах (или крупных
областях) при наличии не менее 5 первичных организаций или 100
участников. Регион координирует работу вузов, организует межвузовские
мероприятия, обеспечивает соблюдение общих стандартов.

**Состав регионального совета:**

-   **Региональный лидер** – избирается советом сроком на один год из
    числа Лидеров первичек или Творцов. Представляет регион в
    Федеральном совете.

-   **Координаторы по лучам** (E, W, J, S, C) – назначаются советом из
    числа Мастеров и выше; помогают кураторам в вузах, организуют
    региональные события по своим направлениям (например, спартакиады,
    эко-акции, конференции).

-   **Координатор по развитию** – отвечает за открытие новых первичных
    организаций, взаимодействие с администрациями вузов, привлечение
    ресурсов.

-   **Координатор по коммуникациям** – ведёт региональные соцсети,
    освещает события, поддерживает связь с федеральным медиа-центром.

-   **Технический координатор** – отвечает за корректное
    функционирование цифровых инструментов в регионе, обучение старост
    работе с таблицами и блокчейном, контроль за ежемесячным
    хешированием данных.

**Функции регионального совета:**

-   Поддержка первичных организаций (методическая, информационная,
    организационная).

-   Проведение региональных Советов (не реже одного раза в квартал) с
    участием Лидеров первичек.

-   Организация региональных слётов, школ актива, конкурсов.

-   Контроль за соблюдением процедур учёта деяний и голосования,
    выборочный аудит данных.

-   Решение спорных вопросов, которые не удалось урегулировать на уровне
    вуза.

-   Подготовка предложений в Федеральный совет по развитию движения.

-   Формирование регионального бюджета (если появится финансирование) и
    распределение ресурсов между вузами.

**Выборы:** Региональный совет формируется на **Региональной
конференции**, куда делегируются представители от каждой первичной
организации (обычно Лидеры) пропорционально численности. Конференция
проходит раз в год, избирает регионального лидера и утверждает
координаторов (по представлению лидера).

5.2.3. Федеральный уровень
--------------------------

Федеральный совет – высший координирующий орган движения в масштабах
страны. Он обеспечивает стратегическое развитие, разрабатывает
стандарты, представляет движение на национальном и международном уровне.

**Состав Федерального совета:**

-   **Председатель** – избирается Федеральным советом из своего состава
    сроком на два года (не более двух сроков подряд). Представляет
    движение официально.

-   **Секретарь** – отвечает за документооборот, протоколы заседаний,
    связь с регионами.

-   **Федеральные кураторы лучей** (5 человек) – избираются из числа
    Творцов и Легенд, имеющих высокий индекс влияния в соответствующем
    луче. Разрабатывают методические материалы, организуют всероссийские
    мероприятия по своим направлениям.

-   **Координатор по развитию** – отвечает за открытие новых регионов,
    взаимодействие с Министерством науки и высшего образования, поиск
    партнёров.

-   **Медиа-координатор** – руководит федеральным медиа-центром,
    обеспечивает единую информационную политику, выпуск новостей,
    ведение сайта и соцсетей.

-   **Технический координатор** – отвечает за цифровую инфраструктуру
    (таблицы, боты, блокчейн), обучение техников в регионах, проведение
    ежемесячного хеширования.

-   **Представители регионов** – в совет могут входить региональные
    лидеры (на ротационной основе или по очереди) для обеспечения
    обратной связи. Оптимальное число – 5–7 человек от разных
    федеральных округов.

**Функции Федерального совета:**

-   Определение стратегии развития движения, утверждение целевых
    показателей (например, охват 5% студентов к определённому году).

-   Разработка и обновление нормативных документов (Устав, Положения о
    статусах, Методика оценки деяний, Правила голосования).

-   Утверждение абсолютных реперов для глобального УЦ и их корректировка
    (если потребуется).

-   Рассмотрение и утверждение экспертных множителей для прорывных
    деяний (на основе заключений внешних экспертов).

-   Организация всероссийских форумов, слётов, конкурсов.

-   Взаимодействие с государственными органами, бизнесом, НКО для
    поддержки движения.

-   Управление федеральным бюджетом (при появлении финансирования).

-   Разрешение конфликтов, которые не удалось решить на региональном
    уровне.

-   Подготовка ежегодного отчёта о деятельности и прогрессе движения
    (глобальный ICD, динамика УЦ и т.п.).

**Выборы:** Федеральный совет формируется на **Всероссийском съезде**,
который собирается раз в два года. Делегаты от регионов (пропорционально
численности участников) избирают Председателя и утверждают кандидатуры
кураторов и координаторов, предложенные Председателем (или выдвинутые
самостоятельно). Съезд также утверждает изменения в Устав и другие
стратегические документы.

5.2.4. Международный уровень (перспектива)
------------------------------------------

При выходе на международный уровень создаётся **Международный
секретариат** (по аналогии с Федеральным советом) и **национальные
комитеты** в странах присутствия. Они действуют на основе общих
принципов, но с учётом местного законодательства и культурных
особенностей. Международный совет координирует глобальные проекты, обмен
опытом, устанавливает единые стандарты оценки (абсолютные реперы) и
может проводить международные слёты.

5.3. Инновационные элементы управления
--------------------------------------

5.3.1. Биократия (репутационное голосование)
--------------------------------------------

В органах управления, где требуется экспертиза (например, при
утверждении множителей для прорывных деяний или выборе кураторов лучей),
вводится **взвешенное голосование** с учётом индекса влияния или ICD
участника. Это позволяет опираться на мнение тех, кто уже доказал свою
компетентность и вклад. Например:

-   При выборе федеральных кураторов лучей каждый делегат съезда имеет
    вес голоса, пропорциональный его индексу влияния (нормированному).

-   При решении спорных вопросов на региональном уровне можно учитывать
    ICD участников, вовлечённых в обсуждение.

Это не отменяет принцип «один человек – один голос» для общих собраний,
но добавляет экспертный слой в специальных вопросах.

5.3.2. Двухпалатная система для стратегических решений
------------------------------------------------------

Для ключевых решений (изменение Устава, установление новых абсолютных
реперов, роспуск движения и т.п.) может применяться двухпалатная
система:

-   **Палата равенства** – все участники движения (один человек – один
    голос) голосуют онлайн.

-   **Палата вклада** – члены Федерального совета, региональные лидеры и
    Легенды (голос пропорционален индексу влияния) голосуют отдельно.  
    Решение принимается, если его поддержали обе палаты
    (квалифицированным большинством). Это защищает от поспешных решений,
    которые могут быть приняты большинством новичков, не понимающих всех
    последствий, а также от узурпации власти элитой.

5.3.3. Проектное управление
---------------------------

Помимо постоянных органов, в движении активно используются **временные
проектные группы** для реализации конкретных инициатив (например,
организация всероссийского форума, разработка мобильного приложения,
проведение масштабной эко-акции). Руководитель проекта назначается
Советом соответствующего уровня из числа Творцов или Легенд, набирает
команду добровольцев и получает необходимые ресурсы. После завершения
проекта группа распускается, а участники получают баллы за деяния.

5.4. Цифровая платформа как основа управления
---------------------------------------------

Вся организационная деятельность опирается на простые и доступные
цифровые инструменты:

-   **Telegram** – основная коммуникационная среда: общие чаты
    (федеральный, региональные, вузовские), каналы для объявлений, боты
    для опросов, голосований, подачи заявок на деяния.

-   **Google Workspace** – таблицы для учёта деяний, голосов, реестра
    участников; формы для заявок; календари мероприятий; диски для
    хранения документов.

-   **Блокчейн-интеграция** – ежемесячная публикация хешей таблиц в сети
    Polygon через мультиподпись, что гарантирует неизменность данных.

-   **Сайт движения** – витрина с информацией о миссии, структуре,
    последних новостях, доске почёта, публичными отчётами (глобальный
    ICD, УЦ).

-   **Веб-интерфейс для голосования** (в перспективе) – может быть
    реализован как отдельное приложение или модуль в сайте, позволяющий
    участникам видеть деяния, голосовать и отслеживать рейтинги.

Цифровые инструменты администрируются техническими координаторами на
каждом уровне. Обучение работе с ними входит в программу наставничества
для новичков.

5.5. Связь структуры со статусной системой
------------------------------------------

Статусы участников (Хранитель, Мастер, Творец, Легенда) напрямую влияют
на возможность занимать выборные должности и участвовать в управлении:

| Статус        | Минимальные требования                              | Права в управлении                                                                                                           |
|---------------|-----------------------------------------------------|------------------------------------------------------------------------------------------------------------------------------|
| **Хранитель** | Любое деяние (≥1 балл)                              | Участие в общих собраниях, голосование в Палате равенства, возможность быть избранным в Совет первички (если есть деяния)    |
| **Мастер**    | 100 баллов в одном луче                             | Право быть старостой, куратором луча в первичке, участвовать в выборах регионального совета, наставничество                  |
| **Творец**    | 300 баллов суммарно + деяния в 3+ лучах             | Право быть избранным в региональный или федеральный совет, руководить проектными группами, участвовать в Палате вклада       |
| **Легенда**   | Индекс влияния \> 10000 (или особое решение Совета) | Пожизненное право участвовать в Палате вклада, статус почётного члена, право вето в чрезвычайных ситуациях (по согласованию) |

5.6. Пример взаимодействия уровней
----------------------------------

Рассмотрим путь инициативы «Всероссийская уборка парков»:

1.  **Идея** возникает у участника (Хранитель) в Томске. Он обсуждает её
    на Совете первички, получает поддержку куратора луча S.

2.  **Первичка** оформляет проектную заявку и направляет её в
    региональный совет (Сибирь).

3.  **Региональный совет** рассматривает, при необходимости уточняет, и
    если проект выходит за рамки региона, передаёт в Федеральный совет с
    рекомендацией.

4.  **Федеральный совет** оценивает масштаб, назначает федерального
    координатора (из кураторов S), выделяет ресурсы (информационная
    поддержка, помощь в поиске спонсоров).

5.  **Проектная группа** формируется из представителей разных регионов,
    руководитель (Творец) координирует подготовку.

6.  **Региональные советы** организуют выезды на местах, обеспечивают
    учёт участников и фиксацию деяний.

7.  После завершения все участники получают баллы (главный луч S,
    дополнительный E), а организаторы – дополнительные баллы за
    организацию (главный луч C или J в зависимости от вклада). Проект
    заносится в реестр, идёт голосование.

8.  Информация о забеге публикуется на сайте, в соцсетях, лучшие фото
    попадают на Доску почёта.

Этот пример иллюстрирует, как структура обеспечивает горизонтальную и
вертикальную координацию, сохраняя при этом гибкость и низкие
бюрократические барьеры.

5.7. Масштабирование и готовность к росту
-----------------------------------------

Предложенная структура легко масштабируется:

-   Новые первичные организации создаются по единому шаблону (достаточно
    найти 5 активистов и провести выборы).

-   Региональные советы формируются автоматически при накоплении
    критической массы.

-   Федеральный совет остаётся компактным, так как регионы представлены
    либо лидерами, либо через механизм делегирования.

-   Цифровые инструменты (таблицы, боты) легко тиражируются и
    поддерживают любое количество участников.

-   Выход на международный уровень потребует создания аналогичных
    структур в других странах с адаптацией под местные законы, но общая
    модель остаётся той же.

5.8. Заключение по разделу
--------------------------

Организационная структура движения «Мы и Наш Дом» сочетает
демократические принципы с элементами меритократии, обеспечивая как
широкое участие, так и эффективное принятие решений. Она опирается на
цифровые инструменты, что делает её прозрачной и гибкой. Статусная
система мотивирует участников расти и брать на себя ответственность, а
двухпалатная модель защищает от перекосов. Всё это позволяет движению
уверенно развиваться, объединяя 5% студентов в каждом вузе и в
перспективе выходя на международный уровень.

Раздел 6. Дорожная карта развития (TRL1–TRL6)
---------------------------------------------

6.1. Общая логика этапов
------------------------

Развитие движения «Мы и Наш Дом» рассматривается как последовательное
прохождение уровней готовности социальной технологии (Social Readiness
Levels), адаптированных под наш проект. Каждый этап завершается
достижением измеримых результатов и готовностью к переходу на следующий
уровень.

| Уровень  | Название              | Краткое описание                                                                                        |
|----------|-----------------------|---------------------------------------------------------------------------------------------------------|
| **TRL1** | Идея                  | Формулировка миссии, базовых принципов, создание инициативной группы                                    |
| **TRL2** | Концепция             | Разработка метрик, процедур учёта, цифровых прототипов, пилот в одной группе                            |
| **TRL3** | Эксперимент           | Проверка модели в контролируемых условиях, сбор обратной связи, корректировка                           |
| **TRL4** | Пилотный вуз          | Развёртывание полного цикла в одном вузе, выборы органов, ежемесячное хеширование                       |
| **TRL5** | Региональный рост     | Масштабирование на несколько вузов, создание региональных советов, проведение слётов                    |
| **TRL6** | Всероссийский масштаб | Достижение охвата 5% студентов в 50+ вузах, устойчивое функционирование, выход на международный уровень |

6.2. Детальное описание этапов
------------------------------

6.2.1. TRL1: Идея (0–3 месяца)
------------------------------

**Цель:** Сформулировать философское ядро и собрать первоначальную
команду.

**Задачи:**

-   **Философия и миссия:** уточнить формулировки пяти лучей, принципы
    «Мы и Наш Дом», проверочный вопрос. Подготовить манифест.

-   **Инициативная группа:** найти 5–7 энтузиастов (из разных вузов или
    одного) для запуска. Определить роли: идеолог, координатор,
    технолог, коммуникатор.

-   **Брендинг:** придумать название, логотип, создать простейший
    сайт-визитку или лендинг.

-   **Коммуникации:** завести Telegram-канал для привлечения первых
    сторонников, опубликовать манифест.

-   **Юридическая консультация:** изучить возможные правовые формы
    (неформальное объединение, общественная организация) для дальнейшей
    регистрации.

**Ключевые результаты:**

-   Пакет идейных документов (манифест, описание пяти лучей).

-   Группа 50+ подписчиков в Telegram, из них 10–15 готовы участвовать в
    пилоте.

-   Зарегистрирован домен и создана страница.

**Ресурсы:**

-   Люди: 5–7 активистов.

-   Инструменты: бесплатные сервисы (Tilda, Telegram, Google Docs).

**Риски и их минимизация:**

-   *Недостаточный интерес:* активный пиар среди знакомых, участие в
    студенческих конференциях.

-   *Размытие идеи:* регулярные встречи инициативной группы, фиксация
    решений.

6.2.2. TRL2: Концепция (3–6 месяцев)
------------------------------------

**Цель:** Детально проработать метрики, процедуры учёта и создать
прототипы цифровых инструментов.

**Задачи:**

-   **Метрики и расчёты:**

    -   Утвердить правила начисления баллов (главный/дополнительные
        лучи).

    -   Разработать алгоритм нормировки и расчёта индивидуального ICD.

    -   Определить шкалу статусов (Мастер, Творец, Легенда) с порогами
        по индексу влияния.

    -   Установить предварительные абсолютные реперы для глобального УЦ
        (на основе экспертных оценок).

-   **Процедуры:**

    -   Описать процесс подачи заявки, валидации старостой, апелляции.

    -   Разработать инструкции для старост и кураторов лучей.

-   **Цифровые прототипы:**

    -   Создать Google Форму для подачи деяний.

    -   Настроить Google Таблицу «Деяния» с автоматическим подсчётом
        сумм и нормировкой (с помощью скриптов Google Apps Script).

    -   Разработать простого Telegram-бота для уведомлений и справочной
        информации (можно через @BotFather).

    -   Протестировать запись хеша таблицы в тестовую сеть Polygon
        (через Remix + MetaMask).

-   **Пилотная группа:** набрать 20–30 добровольцев (можно из одного
    вуза), обучить их, запустить регистрацию деяний на 1–2 месяца.
    Собрать обратную связь.

**Ключевые результаты:**

-   Полный пакет методической документации.

-   Работающий прототип таблицы и бота.

-   Отчёт по пилоту с анализом проблем (сложность заполнения, спорные
    случаи, мотивация).

-   Первые «Хранители» и возможно один «Мастер».

**Ресурсы:**

-   Технический энтузиаст, знакомый с Google Apps Script и основами
    блокчейна.

-   Время участников пилота.

**Риски:**

-   *Сложность интерфейса:* упростить формы, добавить подсказки.

-   *Низкая активность:* проводить еженедельные созвоны, мотивировать
    признанием.

6.2.3. TRL3: Эксперимент (6–12 месяцев)
---------------------------------------

**Цель:** Проверить жизнеспособность модели в контролируемых условиях,
выявить узкие места и скорректировать процедуры.

**Задачи:**

-   **Расширение пилота:** привлечь ещё 50–100 участников (можно из 2–3
    групп разных факультетов). Обеспечить охват всех пяти лучей.

-   **Голосование:** запустить механизм голосования за деяния (сначала
    через бота или простую форму). Настроить таблицу «Голоса»,
    интегрировать расчёт коэффициента $k$.

-   **Блокчейн:** наладить регулярное (раз в месяц) хеширование обеих
    таблиц и публикацию в тестовой сети. Обучить этому одного из
    участников.

-   **Организационная структура:** выбрать временного координатора
    пилота, старост групп. Провести несколько Советов (онлайн или
    оффлайн), записать результаты.

-   **Обратная связь:** провести опросы и интервью с участниками о
    понятности системы, мотивации, справедливости. Выявить частые
    ошибки.

-   **Корректировка:** на основе обратной связи доработать инструкции,
    упростить формы, возможно изменить пороги статусов.

**Ключевые результаты:**

-   Устойчиво работающая система учёта и голосования с ежемесячным
    хешированием.

-   Несколько Советов, зафиксировавших первые интерпретации.

-   Аналитический отчёт с рекомендациями.

-   Первые Творцы (если кто-то набрал \>5000 влияния) – пока
    маловероятно, но возможно.

**Ресурсы:**

-   Техническая поддержка для доработки скриптов.

-   Модераторы для помощи участникам.

**Риски:**

-   *Накрутка голосов:* ввести мониторинг подозрительных паттернов и
    ручной аудит.

-   *Споры об оценке:* разработать простую процедуру апелляции к
    куратору луча.

6.2.4. TRL4: Пилотный вуз (1–2 года)
------------------------------------

**Цель:** Развернуть полную инфраструктуру движения в одном вузе с
охватом не менее 5% студентов (для крупного вуза это 500–1000 человек).

**Задачи:**

-   **Организационное оформление:**

    -   Провести общее собрание, избрать Совет первички (Лидер, старосты
        курсов, кураторы лучей, секретарь).

    -   Разработать и утвердить локальное положение о первичной
        организации.

    -   Начать взаимодействие с администрацией вуза (получить поддержку,
        возможно, помещение для Совета).

-   **Массовое вовлечение:**

    -   Провести презентации на факультетах, в общежитиях.

    -   Запустить информационную кампанию в соцсетях вуза.

    -   Организовать серию открытых мероприятий по пяти лучам
        (спортивные турниры, эко-акции, лекции, концерты).

-   **Цифровая инфраструктура:**

    -   Перевести все таблицы на регулярное автоматическое хеширование
        (возможно, написать скрипт, который раз в месяц сам вычисляет
        хеш и отправляет через мультиподпись).

    -   Обучить технического координатора из числа студентов.

-   **Наставничество:**

    -   Сформировать пул наставников из Мастеров, закрепить за ними
        группы новичков.

    -   Провести обучение наставников.

-   **Система статусов:** начать присваивать статусы автоматически при
    достижении порогов.

**Ключевые результаты:**

-   Действующая первичная организация с выборным советом.

-   Не менее 500 активных участников.

-   Ежемесячные Советы, публичные отчёты.

-   Первые Легенды (если появятся прорывные деяния).

**Ресурсы:**

-   Активисты на всех факультетах.

-   Поддержка администрации (информационная, возможно, небольшая
    финансовая).

-   Технический координатор на 0,25 ставки (или волонтёр).

**Риски:**

-   *Конфликты с администрацией:* строить диалог, подчёркивать пользу
    для репутации вуза и развития студентов.

-   *Перегрузка активистов:* делегировать, ввести ротацию.

6.2.5. TRL5: Региональный рост (2–3 года)
-----------------------------------------

**Цель:** Масштабировать модель на 5–10 вузов в разных регионах, создать
региональные советы и наладить межвузовское взаимодействие.

**Задачи:**

-   **Открытие новых первичек:**

    -   Найти в каждом целевом вузе инициативную группу (можно через
        знакомых, соцсети).

    -   Провести дистанционное обучение (вебинары, передача методичек).

    -   Помочь с первыми выборами и настройкой таблиц.

-   **Региональные советы:**

    -   При достижении в регионе 3–5 первичек провести Региональную
        конференцию, избрать региональный совет.

    -   Определить регионального лидера и координаторов.

-   **Общие мероприятия:**

    -   Организовать межвузовские соревнования по лучам (например,
        спартакиада, экологический квест, научная конференция).

    -   Провести первый региональный слёт актива.

-   **Цифровая платформа:**

    -   Масштабировать таблицы, обеспечить их целостность (можно
        использовать единую таблицу для всего движения с разделением по
        вузам).

    -   Разработать простой дашборд для мониторинга активности по
        регионам.

-   **Юридическое оформление:**

    -   Зарегистрировать межрегиональное общественное движение (или
        выбрать другую форму) для официального статуса.

    -   Открыть расчётный счёт (если появится финансирование).

-   **Фандрайзинг:** начать поиск спонсоров и грантов (например,
    Росмолодёжь, фонды).

**Ключевые результаты:**

-   5–10 первичек, общая численность 2000+ участников.

-   2–3 региональных совета.

-   Проведён региональный слёт.

-   Юридически оформленное движение.

-   Первые поступления от грантов/спонсоров.

**Ресурсы:**

-   Координатор по развитию (федеральный уровень).

-   Юрист-волонтёр или аутсорсинг.

-   Средства на проведение слётов (спонсоры).

**Риски:**

-   *Различие в культуре вузов:* адаптировать методические материалы,
    давать свободу в локальных особенностях.

-   *Сложности координации:* регулярные онлайн-встречи региональных
    лидеров.

6.2.6. TRL6: Всероссийский масштаб (3–5 лет)
--------------------------------------------

**Цель:** Достичь устойчивого охвата 5% студентов в 50+ вузах, создать
полноценную федеральную структуру, обеспечить самофинансирование и
начать международную экспансию.

**Задачи:**

-   **Масштабирование:** продолжить открытие новых первичек, используя
    опыт и методики. К 5 году – 50 вузов, 20 000–30 000 участников.

-   **Федеральные органы:**

    -   Провести Всероссийский съезд, избрать Федеральный совет,
        утвердить устав.

    -   Создать постоянно действующие комитеты (по лучам, по развитию,
        по медиа, по технологиям).

-   **Единая цифровая платформа:**

    -   Разработать мобильное приложение, интегрированное с блокчейном
        (возможно, на базе готового решения с открытым кодом).

    -   Автоматизировать все процессы (подача деяний, голосование,
        расчёт рейтингов, уведомления).

-   **Финансовая устойчивость:**

    -   Создать эндаумент-фонд или привлечь регулярные пожертвования
        выпускников.

    -   Заключить партнёрства с крупными компаниями (заинтересованными в
        подготовке кадров с развитыми soft skills).

-   **Международное направление:**

    -   Перевести основные материалы на английский.

    -   Установить контакты с аналогичными движениями за рубежом
        (например, в рамках программ молодёжных обменов).

    -   Открыть первые пилотные отделения в странах СНГ.

-   **Исследования и развитие:**

    -   Публиковать ежегодные отчёты о прогрессе (глобальный УЦ, анализ
        поколений).

    -   Проводить конференции с участием учёных, обсуждать
        совершенствование метрик.

**Ключевые результаты:**

-   Движение признано в образовательной среде, упоминается в СМИ.

-   Работающее мобильное приложение.

-   Международные контакты и пилоты.

-   Устойчивое финансирование, позволяющее оплачивать работу ключевых
    координаторов (хотя бы частично).

**Ресурсы:**

-   Профессиональная команда: исполнительный директор, разработчики,
    PR-менеджер.

-   Бюджет от грантов, партнёров, членских взносов (символических).

**Риски:**

-   *Бюрократизация:* сохранять горизонтальные связи и проектное
    управление, ротация выборных должностей.

-   *Потеря идеологии:* регулярные Школы актива, Совет старейшин
    (Легенды) как хранители смыслов.

6.6. Заключение по разделу

Дорожная карта обеспечивает поэтапное, контролируемое развитие движения

конкретные задачи, результаты и ресурсы, что позволяет минимизировать
риски и своевременно корректировать курс. Ключевыми факторами успеха
станут вовлечённость активистов, качественная методическая база и
постепенное усиление технической инфраструктуры.

7. Образ будущего
-----------------

7.1. Что такое «образ будущего» для движения «Мы и Наш Дом»?
------------------------------------------------------------

Образ будущего – это не просто прогноз, а **целевой ориентир**, который
вдохновляет участников и задаёт вектор развития. Он основан на нашей
философии, пяти лучах и уровнях цивилизации. Мы строим мир, где каждый
человек осознаёт себя Хозяином – своей жизни, своей планеты, своего
вклада в общее величие. Этот мир не утопия, а достижимое состояние,
которое мы приближаем каждым деянием, каждым голосом, каждым осознанным
выбором.

Образ будущего раскрывается на трёх временных горизонтах: **ближайшее
(10 лет)**, **среднесрочное (20–30 лет)** и **долгосрочное (50+ лет)**.
Каждый горизонт соответствует определённому прогрессу по шкале УЦ и
отражает изменения в жизни участников, общества и планеты.

7.2. Ближайшее будущее: 2035 год – УЦ-2 становится реальностью для движения
---------------------------------------------------------------------------

**Что изменилось в движении:**

-   Движение «Мы и Наш Дом» действует в 100+ вузах России, охватывая
    5–7% студентов (около 200–250 тысяч человек). Первичные организации
    есть в каждом крупном университете, региональные советы работают во
    всех федеральных округах, Федеральный совет признан Министерством
    науки и высшего образования как один из ведущих молодёжных проектов.

-   Более 50 000 участников имеют статус Мастера, 5 000 – Творца, 500 –
    Легенды. Их имена знают не только в вузах, но и за их пределами –
    они становятся лидерами мнений, наставниками, организаторами
    масштабных проектов.

-   Цифровая платформа (мобильное приложение) полностью автоматизировала
    учёт деяний, голосование, расчёт ICD и индекса влияния. Хеши таблиц
    публикуются в блокчейне ежедневно, каждый участник может проверить
    свой прогресс и прогресс друзей.

-   Ежегодный Всероссийский форум «Мы и Наш Дом» собирает тысячи
    участников, на нём вручаются награды Легендам, обсуждаются
    стратегические вопросы, проводятся мастер-классы от выпускников.

**Как изменилась жизнь студентов:**

-   Типичный день студента-участника: утром – зарядка с друзьями (луч
    E), днём – помощь первокурснику с лабораторной (луч W), вечером –
    участие в Совете первички, где он рассказывает, как его курсовой
    проект по биологии может помочь в очистке водоёмов (лучи C и S). Все
    эти деяния автоматически фиксируются в приложении, и он видит, как
    растёт его ICD и индекс влияния.

-   На доске почёта в холле университета – фото студентов с высокими
    показателями. Их портреты сопровождаются не только учебными
    достижениями, но и вкладом в пять лучей. Работодатели, которые знают
    о движении, запрашивают у выпускников не только диплом, но и
    цифровой профиль KIT – это становится дополнительным преимуществом
    при трудоустройстве.

-   В учебные программы некоторых вузов начинают включать факультативы
    по «философии Мы и Наш Дом» и методике интерпретации научной
    деятельности в терминах ICD.

**Вклад в глобальный прогресс:**

-   Благодаря активности движения, максимумы исходных баллов по
    некоторым лучам (особенно S – экология и C – культура) значительно
    выросли. Глобальный УЦ, рассчитываемый на основе этих максимумов,
    поднялся с 0,0005 до 0,001 (при тех же абсолютных реперах). Хотя это
    всё ещё далеко от УЦ-2, тренд обнадёживает.

-   Начинается постепенный переход к УЦ-2 в масштабах страны: благодаря
    тысячам волонтёрских и экологических акций, а также научным
    разработкам участников, в отдельных регионах базовые блага
    становятся доступнее. Например, в пилотных областях запускаются
    программы «Доступная среда» с участием наших активистов.

7.3. Среднесрочное будущее: 2040–2050 годы – от УЦ-2 к УЦ-3
-----------------------------------------------------------

**Мир вокруг:**

-   Человечество в целом приближается к УЦ-2 (глобальный ICD по нашей
    методике достигает 0,4–0,5). Это означает, что в развитых странах
    базовые блага гарантированы всем, а в развивающихся – стремительно
    распространяются благодаря новым технологиям и социальным
    инициативам.

-   Энергетика становится преимущественно возобновляемой и термоядерной.
    Появляются первые коммерческие термоядерные реакторы (возможно,
    созданные участником движения, как в нашем примере). Это резко
    поднимает максимум по лучу E и, соответственно, глобальный УЦ.

-   Экологические программы приводят к восстановлению лесов и снижению
    углеродного следа. Луч S растёт благодаря тысячам локальных акций и
    глобальным соглашениям.

**Движение «Мы и Наш Дом»:**

-   Действует не только в России, но и в 20+ странах (СНГ, Европа,
    Азия). Создан Международный секретариат, национальные комитеты
    адаптируют методику к местным условиям.

-   Количество участников превышает 5 миллионов, из них около 300 тысяч
    – Мастера, 30 тысяч – Творцы, 3 тысячи – Легенды. Статус Легенды
    теперь требует не только высокого индекса влияния, но и
    подтверждённого вклада в глобальные прорывы (например, научные
    открытия, социальные реформы).

-   Цифровая платформа стала многоязычной, интегрирована с
    международными базами научных публикаций и патентов, что позволяет
    автоматически учитывать деяния (с согласия участников).
    Блокчейн-инфраструктура использует несколько сетей для повышения
    надёжности.

-   Выпускники движения занимают ключевые позиции в науке, бизнесе,
    государственном управлении. Они применяют принципы пяти лучей в
    своей работе, создавая «Мы и Наш Дом-ориентированные» компании и
    проекты.

**Жизнь участника (2045 год):**

-   Молодой учёный, участник с детства (вступил в 2030 году), защищает
    диссертацию по биомимикрии. Его работа уже в процессе приносит баллы
    по лучам C и S. Он получает статус Творца и приглашение войти в
    международный экспертный совет по устойчивому развитию.

-   На Совете своего региона он предлагает проект глобальной сети
    «зелёных» коридоров, который получает поддержку и финансирование из
    международного фонда движения.

-   Его цифровой профиль KIT содержит полную историю деяний, включая
    школьные экологические акции, и служит основой для репутации в
    профессиональном сообществе.

**Глобальный УЦ:**

-   Благодаря прорывам в энергетике и экологии, а также росту
    справедливости (благодаря социальным проектам), глобальный УЦ
    достигает 0,6–0,7. Человечество входит в эру УЦ-3: энергия звезды
    становится доступной, начинается активное освоение ближнего космоса
    (лунные базы, орбитальные станции).

7.4. Долгосрочное будущее: 2075–2100 годы – гармония и величие (УЦ-4 и выше)
----------------------------------------------------------------------------

**Человечество как единый организм:**

-   Планета превращается в сеть осознанных сообществ, объединённых
    общими ценностями пяти лучей. Национальные границы теряют былое
    значение, уступая место культурному разнообразию внутри единой
    цивилизации.

-   Энергетика полностью безуглеродная, термоядерный синтез и
    космическая энергетика обеспечивают изобилие. Ресурсы
    перераспределяются так, что каждый человек имеет доступ к базовым
    благам (УЦ-2 давно достигнут, идёт движение к УЦ-4).

-   Экология восстановлена: леса, океаны, биоразнообразие – под защитой,
    а города вписаны в природные ландшафты. Луч S близок к абсолютному
    реперу.

-   Культура и наука переживают ренессанс: творчество становится
    основной формой самореализации, научные открытия делаются ежедневно
    благодаря коллективному разуму, поддерживаемому цифровыми
    платформами и системой наставничества.

**Движение «Мы и Наш Дом»:**

-   Превратилось в глобальную сеть, объединяющую миллиарды людей (хотя
    формально участие остаётся добровольным). Многие даже не помнят,
    когда вступили – это стало естественной частью взросления.

-   Статус Легенды теперь присваивается не только живущим, но и
    посмертно – тем, чей вклад продолжает влиять на поколения. Их
    «цифровые души» (токены KIT) хранятся в мемориальных
    блокчейн-реестрах.

-   Система оценки постоянно совершенствуется, но базовые принципы
    остаются неизменными, обеспечивая преемственность поколений.

**Человек будущего (2080 год):**

-   Ребёнок с ранних лет знакомится с пятью лучами через игровые
    симуляторы. Его первый «вклад» – уход за домашним растением (луч S)
    – фиксируется в семейном блокчейн-приложении.

-   Став студентом, он уже имеет начальный ICD и понимает, как его
    увлечения и учёба связаны с глобальным прогрессом. Он выбирает
    проектную деятельность, которая резонирует с его сильными лучами, и
    получает поддержку от сообщества.

-   В зрелом возрасте он сам становится наставником, помогая новым
    поколениям находить своё место в мире пяти лучей.

**Космическая экспансия:**

-   Человечество осваивает Солнечную систему. На Марсе и Луне действуют
    научные станции, где участники движения реализуют проекты по
    терраформированию, используя принципы устойчивости (S) и энергии
    (E). Каждое такое достижение увеличивает глобальные максимумы,
    приближая нас к УЦ-4 и выше.

7.5. Что остаётся неизменным
----------------------------

Несмотря на все технологические и социальные изменения, **философское
ядро** движения остаётся незыблемым:

-   Человек – Хозяин, а не арендатор.

-   Достаточно просто быть, но быть осознанно.

-   Пять лучей – универсальный компас.

-   Вопрос «Правда ли мы то самое человечество, которое может?»
    продолжает звучать на Советах, напоминая, что величие требует
    постоянного подтверждения.

7.6. Заключительная мысль
-------------------------

Образ будущего, который мы строим, – это не далёкая мечта, а **проекция
сегодняшних деяний**. Каждое участие в забеге, каждая уборка территории,
каждая научная статья, каждый акт доброты и справедливости – это

Через десять лет мы увидим первые плоды: тысячи выпускников с высоким
ICD, сотни социальных и технологических проектов, зарождающуюся культуру
гармоничного развития. Через тридцать лет наши дети будут воспринимать
пять лучей как нечто естественное, как таблицу умножения. А через
пятьдесят лет человечество, возможно, оглянется и скажет: «Мы сделали
это. Мы стали Мы и Наш Домми».

**«Правда ли мы то самое человечество, которое может?» – мы отвечаем

ИССЛЕДОВАНИЯ
============

1. Уточнение абсолютных реперов
-------------------------------

1.1. Постановка проблемы
------------------------

Абсолютные реперы $L_{l}^{\text{abs}}$ являются фундаментом глобального
Уровня Цивилизации (УЦ). Они задают «точку отсчёта», относительно
которой измеряется прогресс человечества. В текущей версии модели реперы
определены гипотетически (например, $L_{E}^{\text{abs}}$ соответствует
полной мощности солнечного излучения). Однако для практического
использования и обеспечения доверия к системе необходима их **научная
валидация** и, возможно, корректировка на основе международных
стандартов и экспертных оценок.

1.2. Обзор существующих подходов к измерению цивилизационного развития
----------------------------------------------------------------------

Анализ существующих методов позволяет выделить несколько подходов,
которые могут быть использованы для обоснования абсолютных реперов:

1.2.1. Индекс социального развития Морриса
------------------------------------------

Историк Иэн Моррис предложил количественный метод измерения прогресса
цивилизаций на основе четырёх компонентов: энергопотребление,
организация общества (максимальный размер поселения), информационные
технологии и военный потенциал . Он использовал **индексный метод** с
фиксированной точкой отсчёта (2000 год = 1000 баллов) и распределением
весов между компонентами. Этот подход демонстрирует возможность
построения долгосрочных шкал развития, но его реперы привязаны к
конкретному историческому моменту, а не к абсолютным пределам.

1.2.2. Шкала Кардашёва
----------------------

Исходная шкала Кардашёва (Тип I, II, III) основана на логарифмической
зависимости от энергопотребления . Она даёт естественные абсолютные
реперы: энергия, получаемая планетой от звезды (Тип I), полная мощность
звезды (Тип II), мощность галактики (Тип III). Эти значения могут быть
рассчитаны с высокой точностью на основе астрофизических данных.

1.2.3. Индексы устойчивого развития ООН (ЦУР)
---------------------------------------------

Цели устойчивого развития ООН содержат множество конкретных измеримых
показателей, которые могут служить основой для определения реперов по
лучам W, J, S. Например:

-   Отсутствие голода, чистые вода и санитария (W).

-   Сокращение неравенства, мир и правосудие (J).

-   Борьба с изменением климата, сохранение экосистем (S).

1.2.4. Индекс человеческого развития (ИЧР)
------------------------------------------

ИЧР комбинирует три показателя: ожидаемая продолжительность жизни,
уровень образования и валовой национальный доход на душу населения. Он
даёт пример многомерной нормировки с использованием минимальных и
максимальных значений (например, для дохода используется логарифмическая
шкала с порогами).

1.3. Методология уточнения абсолютных реперов
---------------------------------------------

Предлагается **трёхэтапная процедура** уточнения реперов, сочетающая
экспертные оценки, анализ международных стандартов и математическое
моделирование.

1.3.1. Этап 1: Сбор и анализ существующих ориентиров
----------------------------------------------------

Для каждого луча составляется список потенциальных «кандидатов» в
абсолютные реперы из различных источников:

| Луч   | Возможные источники реперов                                                                                                   |
|-------|-------------------------------------------------------------------------------------------------------------------------------|
| **E** | Шкала Кардашёва (энергия Типа I, II), текущее мировое энергопотребление, прогнозы развития термоядерной энергетики            |
| **W** | Показатели ЦУР (отсутствие голода, доступ к воде), прожиточный минимум в развитых странах, концепция «базового дохода»        |
| **J** | Индекс Джини (целевое значение 0), Индекс восприятия коррупции, показатели гендерного равенства                               |
| **S** | Целевые показатели Парижского соглашения (нулевые выбросы к 2050–2100), площадь лесов, биоразнообразие (индекс живой планеты) |
| **C** | Доступ к интернету, индекс образования, количество научных публикаций, Индекс свободы творчества (где существует)             |

1.3.2. Этап 2: Экспертная оценка (метод Дельфи)
-----------------------------------------------

Проводится многораундовое анкетирование экспертов по следующим вопросам:

-   Является ли данный репер **достижимым** для человечества в обозримом
    будущем?

-   Является ли он **желательным** с этической точки зрения?

-   Отражает ли он **максимально возможный** уровень развития, или это
    лишь промежуточная цель?

-   Какова должна быть **динамика** (скорость) приближения к реперу?

В качестве экспертов привлекаются:

-   Учёные (физики, экологи, экономисты, социологи).

-   Представители международных организаций (ООН, ЮНЕСКО, Всемирный
    банк).

-   Лидеры движения «Мы и Наш Дом» (особенно Легенды и Творцы с высоким
    индексом влияния в соответствующем луче).

-   Футурологи и специалисты по долгосрочному прогнозированию.

1.3.3. Этап 3: Математическое обоснование
-----------------------------------------

Для каждого луча выбирается **шкала** (линейная, логарифмическая,
степенная), наилучшим образом отражающая прогресс. Например, для энергии
(E) очевидно подходит логарифмическая шкала (как у Кардашёва), поскольку
переход от 1 ТВт к 2 ТВт качественно менее значим, чем от $10^{16}$ Вт
к $10^{17}$ Вт.

Для лучей W, J, S, C может использоваться **комбинированная шкала**,
сочетающая линейные отрезки с насыщением (например, обеспеченность
базовыми благами может измеряться линейно от 0% до 100%, а после 100% –
переход к качественно новым показателям).

1.4. Предварительные предложения по абсолютным реперам
------------------------------------------------------

На основе предварительного анализа предлагаются следующие кандидаты
(подлежат уточнению через экспертные процедуры):

| Луч   | Абсолютный репер $L^{\text{abs}}$                                                                  | Обоснование                                                                                                                                              |
|-------|----------------------------------------------------------------------------------------------------|----------------------------------------------------------------------------------------------------------------------------------------------------------|
| **E** | $3,8 \times 10^{26}$ Вт                                                                            | Полная мощность Солнца – энергетический потенциал звезды, доступный для цивилизации Типа II по Кардашёву.                                                |
| **W** | 100% обеспечение базовых потребностей (питание, вода, жильё, медицина) для всех людей на планете   | Цель, зафиксированная во Всеобщей декларации прав человека и ЦУР 1,2,3,6.                                                                                |
| **J** | Индекс Джини = 0 (абсолютное равенство)                                                            | Идеал справедливости, к которому стремится человечество (хотя практически недостижим). Может быть заменён на более реалистичный, например, Джини \< 0,2. |
| **S** | Нулевой чистый углеродный след + восстановление экосистем до уровня 1850 года                      | Цели Парижского соглашения и программы ООН по восстановлению экосистем.                                                                                  |
| **C** | Доступ ко всем знаниям человечества для каждого человека + свобода творчества (индекс свободы = 1) | Идеал просвещения: каждый может учиться, творить и делиться результатами.                                                                                |

1.5. Механизм обновления реперов
--------------------------------

Поскольку абсолютные реперы должны оставаться неизменными для сохранения
сравнимости во времени, их изменение – крайне редкое событие.
Предлагается следующий протокол:

1.  **Инициация пересмотра:** не чаще одного раза в 10 лет (или по
    достижении глобальным УЦ значения \>0,9).

2.  **Экспертный совет:** формируется из 50–100 ведущих специалистов
    мира (по рекомендациям научных организаций).

3.  **Открытое обсуждение:** проект новых реперов публикуется на срок не
    менее 6 месяцев для сбора отзывов от участников движения и всех
    заинтересованных лиц.

4.  **Голосование:** решение о новых реперах принимается двухпалатной
    системой движения (Палата равенства и Палата вклада)
    квалифицированным большинством (2/3).

5.  **Историческая ретроспектива:** при изменении реперов все предыдущие
    значения глобального УЦ пересчитываются по новой шкале для
    сохранения преемственности.

1.6. Пример уточнения репера для луча S (Экология)
--------------------------------------------------

Рассмотрим возможный процесс уточнения для луча S.

**Исходное предположение:** $L_{S}^{\text{abs}} =$ «нулевой углеродный
след + восстановление экосистем до уровня 1850 года».

**Этап 1 (сбор данных):**

-   Парижское соглашение: цель – достичь нулевых выбросов к 2050–2100
    году.

-   Программа ООН «Десятилетие восстановления экосистем» (2021–2030):
    конкретные показатели по площади восстановленных лесов, торфяников и
    т.д.

-   Индекс живой планеты (WWF): показывает изменение биоразнообразия
    (1970 = 1, сейчас \~0,6).

**Этап 2 (экспертная оценка):**  
Опрос климатологов, экологов, экономистов:

-   Достижимо ли восстановление до уровня 1850 года? (Многие скажут, что
    нет, из-за необратимых изменений).

-   Может быть, лучше взять за репер «стабилизацию климата на уровне
    +1,5°C» и «сохранение 30% территории в естественном состоянии» (цели
    Конвенции о биоразнообразии)?

**Этап 3 (математическое обоснование):**  
Построение индекса, объединяющего несколько показателей:

$$y_{S} = \frac{1}{3}\left( \frac{CO_{2,current}}{CO_{2,1850}}+\frac{B_{\text{current}}}{B_{1850}}+\frac{F_{\text{current}}}{F_{\text{target}}} \right)
$$

где $CO_{2}$ – концентрация парниковых газов, $B$ – индекс
биоразнообразия, $F$ – доля охраняемых территорий.

**Результат:** Абсолютный репер может стать составным и более
реалистичным, но при этом сохранить свою функцию «абсолютной цели».

1.7. Ожидаемые результаты исследования
--------------------------------------

1.  **Научно обоснованные абсолютные реперы** по каждому из пяти лучей,
    одобренные экспертным сообществом.

2.  **Методика пересчёта** исторических данных для обеспечения
    преемственности.

3.  **Открытая база данных** всех использованных источников и расчётов.

4.  **Периодически обновляемые доклады** о прогрессе человечества по
    шкале УЦ, публикуемые движением.

1.8. Связь с другими направлениями исследований
-----------------------------------------------

Уточнение абсолютных реперов неразрывно связано с:

-   **Развитием методики сбора данных** (п.3) – чем точнее мы измеряем
    текущие максимумы, тем надёжнее сравнение с реперами.

-   **Исследованием корреляции ICD и реальных достижений** – необходимо
    проверять, действительно ли рост максимумов ведёт к улучшению
    качества жизни.

-   **Анализом поколенческой динамики** – как разные поколения
    приближаются к реперам.

1.9. Заключение
---------------

Уточнение абсолютных реперов – ключевое научное направление,
обеспечивающее долгосрочную устойчивость и доверие к системе ICD. Оно
требует междисциплинарного подхода, сочетающего естественнонаучные,
социальные и гуманитарные знания. Движение «Мы и Наш Дом» может стать
уникальной площадкой, где эти исследования проводятся с участием
заинтересованных студентов и молодых учёных под руководством экспертов,
формируя новое поколение исследователей, мыслящих в терминах
планетарного развития.

2. Изучение возможности включения дополнительных лучей (например, «Технологии» или «Инновации») без потери простоты
-------------------------------------------------------------------------------------------------------------------

2.1. Актуальность вопроса
-------------------------

Пять лучей (E, W, J, S, C) были выбраны как минимальный набор,
охватывающий основные измерения человеческого развития. Однако по мере
развития движения и углубления понимания феномена «величие человечества»
может возникнуть потребность в расширении этого набора. В частности,
часто обсуждаются такие кандидаты, как:

-   **Технологии (T)** – уровень технологического развития, изобретения,
    инженерные решения.

-   **Инновации (I)** – способность создавать новое, внедрять улучшения.

-   **Цифровизация (D)** – доступ к цифровым технологиям, уровень
    автоматизации.

-   **Безопасность (Sec)** – физическая и кибербезопасность,
    устойчивость к угрозам.

-   **Духовность (Sp)** – глубже, чем культура, возможно, отдельное
    измерение.

Включение нового луча должно быть обосновано и, главное, не должно
нарушать простоту и прозрачность системы, которая является одним из
ключевых принципов движения.

2.2. Анализ потенциальных лучей
-------------------------------

2.2.1. Технологии (T)
---------------------

**Что измеряет:** уровень технологического развития, количество
патентов, научно-технические достижения, внедрение инноваций в
производство и быт.  
**Пересечение с существующими лучами:**

-   Технологии тесно связаны с **C (Культура)** – наука и знания, а
    также с **E (Энергия)** – энергоэффективные технологии.

-   Создание новых технологий часто требует творчества (C) и влияет на
    благосостояние (W).  
    **Вывод:** Технологии могут рассматриваться как подкатегория C или
    как отдельный луч, если мы хотим подчеркнуть их самостоятельную
    ценность.

2.2.2. Инновации (I)
--------------------

**Что измеряет:** способность создавать и внедрять новое, стартапы,
инновационная активность.  
**Пересечение:** близко к технологиям, но шире (включает социальные
инновации, организационные). Может быть частью C и J (социальные
инновации для справедливости).

2.2.3. Цифровизация (D)
-----------------------

**Что измеряет:** доступ к интернету, цифровые навыки, автоматизация.  
**Пересечение:** в современном мире это уже часть культуры (C) и блага
(W) (доступ к информации). Может быть включено в существующие лучи.

2.2.4. Безопасность (Sec)
-------------------------

**Что измеряет:** защищённость от преступности, военных конфликтов,
киберугроз, экологических катастроф.  
**Пересечение:** частично отражается в J (справедливость, правопорядок)
и S (экологическая безопасность). Но может быть и отдельным аспектом.

2.2.5. Духовность (Sp)
----------------------

**Что измеряет:** внутренний мир, этические ценности, религиозные или
философские убеждения.  
**Пересечение:** близко к C (культура), но уже. В рамках движения мы уже
заложили духовное развитие в философию «Мы и Наш Дом», и культура (C)
включает творчество и знания, но не обязательно духовные практики.

2.3. Критерии необходимости добавления луча
-------------------------------------------

Прежде чем вводить новый луч, следует ответить на вопросы:

1.  **Охватывает ли существующий набор данное измерение?** Если да, то
    добавление избыточно.

2.  **Является ли это измерение фундаментальным и универсальным** для
    всех людей и культур?

3.  **Можно ли его измерить** с той же степенью объективности и
    простоты, как существующие лучи?

4.  **Не приведёт ли добавление к усложнению** системы до такой степени,
    что она станет непонятной для массового участника?

5.  **Будет ли это стимулировать развитие** в новом направлении или
    просто добавит баллы за то, что и так учитывается?

2.4. Риски добавления новых лучей
---------------------------------

-   **Усложнение расчёта:** среднее геометрическое шести величин менее
    наглядно, чем пяти. Каждый новый луч требует пересмотра нормировки и
    может сделать ICD менее чувствительным к изменениям.

-   **Дробление смыслов:** вместо целостной картины участники могут
    начать «гоняться» за отдельными лучами, теряя гармонию.

-   **Проблема весов:** придётся решать, равны ли лучи по значимости,
    или вводить веса, что добавит субъективности.

-   **Увеличение бюрократии:** нужно будет классифицировать деяния по
    большему числу категорий, что повысит нагрузку на валидаторов.

2.5. Способы интеграции новых аспектов без потери простоты
----------------------------------------------------------

2.5.1. Подкатегории внутри существующих лучей
---------------------------------------------

Самый простой способ – детализировать существующие лучи, добавив к ним
подпункты. Например, луч C (Культура) можно расширить, включив в него
«научно-техническое творчество», «инновации», «цифровые компетенции».
Это не меняет структуру индекса, но позволяет точнее классифицировать
деяния. В описании луча можно дать более широкие примеры.

2.5.2. Специализированные индексы для отдельных групп
-----------------------------------------------------

Для исследовательских целей или для оценки определённых проектов можно
вводить временные или локальные дополнительные показатели, которые не
влияют на основной ICD. Например, в рамках конкурса инновационных
проектов можно использовать дополнительный индекс инновационности,
который суммируется отдельно.

2.5.3. Включение в индекс влияния
---------------------------------

Новые аспекты (например, технологические достижения) могут учитываться
через **индекс влияния**, если за них голосуют участники. При этом
объективная ценность (исходные баллы) может оставаться в рамках
существующих лучей (например, патент – это C, а его влияние на
сообщество отразится в голосах). Это уже работает в текущей модели.

2.5.4. Периодический пересмотр пяти лучей
-----------------------------------------

Движение может проводить раз в несколько лет широкое обсуждение и,
возможно, корректировать набор лучей, если накопится достаточно
аргументов. Например, через 10 лет, когда технологии станут ещё важнее,
можно рассмотреть замену какого-то луча на более актуальный или
добавление шестого. При этом важно сохранить преемственность: старые
данные должны быть пересчитаны.

2.6. Исследовательские задачи
-----------------------------

Для принятия обоснованного решения необходимо провести следующие
исследования:

1.  **Опрос участников движения** (и потенциальных участников) о том,
    считают ли они, что пять лучей охватывают все важные аспекты их
    жизни и развития. Какие аспекты, по их мнению, упущены?

2.  **Анализ существующих международных индексов** (например, Глобальный
    инновационный индекс, Индекс готовности к цифровой экономике) и их
    пересечений с нашими лучами.

3.  **Пилотное тестирование** дополнительных лучей в нескольких
    первичных организациях: предложить участникам классифицировать свои
    деяния также по новым лучам (например, T) и посмотреть, насколько
    часто они будут использоваться и не дублируют ли существующие.

4.  **Статистический анализ корреляции** между новыми кандидатами и
    существующими лучами. Если корреляция очень высока
    (например, \>0.9), то новый луч не добавляет информации.

5.  **Моделирование влияния** добавления луча на распределение ICD и
    статусов среди участников.

2.7. Предварительные рекомендации
---------------------------------

На текущем этапе (первые 5–10 лет развития движения) **не рекомендуется
вводить новые основные лучи**. Существующие пять (E, W, J, S, C)
доказали свою полноту и простоту. Они достаточно широки, чтобы охватить
технологии и инновации через подкатегории. Например:

-   **E** может включать энергоэффективные технологии.

-   **C** может включать научные открытия, изобретения, инновации.

-   **J** может включать социальные инновации, направленные на
    справедливость.

-   **W** может включать технологии, улучшающие благосостояние
    (медицина, сельское хозяйство).

-   **S** может включать «зелёные» технологии.

Такая трактовка уже заложена в примерах: «обработка изображений поможет
сделать конвейер по сортировке мусора» – это интерпретируется как вклад
в S и C.

Однако в долгосрочной перспективе, когда движение станет глобальным и
появятся новые вызовы (например, освоение космоса, искусственный
интеллект), возможно, потребуется выделить, скажем, «Космический» луч
или «Цифровой». К этому времени накопится достаточно данных и опыта,
чтобы принять взвешенное решение.

2.8. Заключение
---------------

Изучение возможности включения дополнительных лучей – важное
исследовательское направление, обеспечивающее эволюцию системы. Оно
должно проводиться регулярно, с участием широкого круга заинтересованных
сторон. Приоритетом всегда остаётся сохранение простоты и понятности для
участников. Наиболее гибким и безопасным способом учёта новых аспектов
является их интеграция в существующие лучи через расширение трактовок и
использование индекса влияния, а не добавление новых размерностей в
основной индекс гармоничности.

3. Разработка методики перевода реальных достижений в баллы системы
-------------------------------------------------------------------

3.1. Актуальность и общая постановка задачи
-------------------------------------------

В текущей версии системы участники регистрируют свои деяния через
интерпретацию в терминах пяти лучей. Однако многие реальные достижения
имеют объективную, внешнюю шкалу оценки: спортивные разряды (кандидат в
мастера спорта, мастер спорта), научные степени (кандидат наук, доктор
наук), волонтёрские часы (подтверждённые на портале Добро.рф), победы в
олимпиадах, публикации в рецензируемых журналах и т.д. Эти достижения
уже прошли независимую экспертизу и могут быть напрямую конвертированы в
баллы системы, что повысит объективность и снизит нагрузку на
валидаторов.

Методика перевода должна удовлетворять следующим требованиям:

-   **Прозрачность** – правила перевода должны быть чёткими и доступными
    для понимания.

-   **Справедливость** – эквивалентные по сложности достижения в разных
    областях должны давать сопоставимое количество баллов.

-   **Гибкость** – возможность адаптировать шкалы под новые виды
    достижений.

-   **Связь с философией пяти лучей** – каждое достижение должно быть
    отнесено к одному или нескольким лучам.

3.2. Теоретические основы перевода
----------------------------------

Перевод результатов из одной шкалы в другую – классическая задача теории
оценок . Если тесты или достижения измеряются в разных единицах
(секунды, килограммы, часы, баллы ЕГЭ), прямое сравнение абсолютных
значений невозможно. Необходимо использовать унифицированную меру
– **баллы системы ICD** .

Процесс оценивания включает три стадии :

1.  Подбор шкалы для перевода результатов в баллы.

2.  Преобразование результатов в баллы в соответствии с выбранной
    шкалой.

3.  Сравнение полученных баллов с нормами (в нашем случае – учёт в общем
    рейтинге участника).

3.3. Обзор существующих подходов
--------------------------------

3.3.1. Вузовская система учёта индивидуальных достижений
--------------------------------------------------------

Многие вузы при приёме абитуриентов используют баллы за индивидуальные
достижения . Например, НГУЭУ начисляет:

-   Золотой знак ГТО – 3 балла, серебряный – 2, бронзовый – 1 .

-   Волонтёрская деятельность: 61+ часов – 3 балла, 41–60 часов – 2
    балла, до 40 часов – 1 балл .

-   Документ об образовании с отличием – 10 баллов .

-   Победа в олимпиадах – 2–4 балла .

Этот подход отличается простотой, но не всегда учитывает качественные
различия между достижениями одного уровня (например, победа в разных
олимпиадах оценивается одинаково).

3.3.2. Система поощрения волонтёров
-----------------------------------

В Свердловской области действует «Магазин поощрений» для волонтёров, где
часы конвертируются в баллы с повышающим коэффициентом в зависимости от
уровня :

-   Базовый уровень (0–149 часов): конвертация 1:1.

-   Бронзовый уровень (150–299 часов): 1:1,5.

-   Золотой уровень (300+ часов): 1:2 .

Такой подход стимулирует долгосрочное участие и может быть адаптирован
для нашего движения.

3.3.3. Спортивная классификация (ЕВСК)
--------------------------------------

В Единой всероссийской спортивной классификации используется **принцип
эквивалентности результатов** . Для многоборий рассчитывается
«коэффициент реализации» (КР) – отношение результата в многоборье к
сумме очков, оценивающих эквивалентные результаты в отдельных видах.
Например, для женского семиборья КР составляет 0,78–0,82 . Это позволяет
объективно сравнивать достижения в разных дисциплинах.

3.3.4. Статистические методы перевода тестовых баллов
-----------------------------------------------------

В работах по переводу результатов тестирования предлагаются
математические методы, учитывающие распределение результатов . Например,
на основе предположения о нормальном распределении внутри групп можно
объективно выделить границы оценок, минимизируя субъективный фактор .
Эти методы могут быть применены для калибровки шкал перевода.

3.4. Предлагаемая методика перевода
-----------------------------------

3.4.1. Общие принципы
---------------------

1.  **Каждое достижение привязывается к одному или нескольким
    лучам** (например, спортивный разряд – к E, научная степень – к C,
    волонтёрские часы – к W).

2.  **Используется единая базовая шкала баллов**, относительно которой
    выравниваются разные виды достижений. За базовую единицу
    предлагается принять **1 балл ICD**, эквивалентный одному часу
    волонтёрской деятельности базового уровня (это легко проверить и
    имеет понятную размерность).

3.  **Для каждого вида достижений разрабатывается таблица
    соответствия** с учётом экспертных оценок и, при наличии,
    статистических данных о распределении.

4.  **Периодическая калибровка** – раз в 2–3 года шкалы могут уточняться
    на основе накопленных данных.

|       | Победа в олимпиаде                   | Всероссийская (призёр)     | 50           | Аналогично вузовским баллам                                           |
|       |                                      | Международная (победитель) | 200          |                                                                       |

3.4.3. Учёт множителей и комбинаций
-----------------------------------

-   Если достижение относится к нескольким лучам, баллы могут
    суммироваться (например, научная статья по экологичным технологиям –
    C и S), но с понижающим коэффициентом (например, 0,7), чтобы
    избежать двойного счёта.

-   Для уникальных достижений (Нобелевская премия, олимпийское золото)
    может применяться экспертный множитель (как для прорывных деяний).

3.5. Метод калибровки шкал на основе статистического подхода
------------------------------------------------------------

Для объективного установления соответствия между различными видами
достижений предлагается использовать метод, аналогичный описанному
в (<https://journals.rcsi.science/1816-9228/article/view/253713>) , но
адаптированный для нашей задачи.

Пусть у нас есть два вида достижений A и B (например, спортивные разряды
и волонтёрские часы). Мы имеем выборку участников, у которых есть оба
типа достижений. Тогда:

1.  Строим распределение участников по каждому виду достижений.

2.  Используя метод максимизации правдоподобия, подбираем функцию
    перевода f(A) = B, минимизирующую расхождение между фактическими и
    ожидаемыми значениями.

3.  Для калибровки привлекаются эксперты, которые задают начальные
    приближения (например, КМС ≈ 100 часов волонтёрства).

Этот подход позволит сделать шкалы самонастраивающимися по мере
накопления данных.

3.6. Процедура внедрения методики
---------------------------------

1.  **Создание рабочей группы** из экспертов по каждому лучу (спортивные
    тренеры, учёные, лидеры волонтёрских организаций).

2.  **Разработка первоначальных таблиц соответствия** на основе анализа
    существующих систем .

3.  **Пилотное тестирование** в нескольких первичных организациях:
    участники предоставляют подтверждённые внешние достижения, которые
    переводятся в баллы по новой методике, и сравниваются с их текущими
    баллами.

4.  **Сбор обратной связи** и корректировка шкал.

5.  **Утверждение методики** Федеральным советом и публикация в открытом
    доступе.

6.  **Интеграция в цифровую платформу**: добавление в форму подачи
    деяния возможности указать внешнее достижение и автоматический
    расчёт баллов (с возможностью ручной корректировки).

3.7. Пример расчёта
-------------------

**Участница Анна:**

-   Имеет золотой знак ГТО (луч E) – 20 баллов.

-   Отработала 200 волонтёрских часов (луч W). Базовый расчёт: 200 часов
    × 1,5 (бронзовый уровень) = 300 баллов.

-   Опубликовала статью в журнале ВАК (луч C) – 100 баллов.

Итого дополнительных баллов: 20 + 300 + 100 = 420. Эти баллы добавляются
к её исходным баллам (не путать с взвешенными баллами от голосования).
Они учитываются при расчёте максимумов и, соответственно, влияют на
глобальный УЦ, так как повышают объективные показатели развития.

3.8. Заключение
---------------

Разработанная методика перевода реальных достижений в баллы системы ICD
позволяет:

-   объективно учитывать внешне подтверждённые заслуги участников;

-   снизить нагрузку на валидаторов;

-   повысить мотивацию за счёт признания достижений, уже полученных в
    официальных структурах;

-   обеспечить сопоставимость результатов в разных областях.

Дальнейшие исследования должны быть направлены на уточнение
коэффициентов эквивалентности с использованием статистических методов и
расширение перечня учитываемых достижений.

4. Исследование корреляции между ICD и реальными жизненными успехами участников
-------------------------------------------------------------------------------

4.1. Актуальность и цель исследования
-------------------------------------

Индекс Цивилизационного Развития (ICD) призван измерять гармоничность
развития личности и её социальное влияние. Однако для подтверждения его
валидности и практической значимости необходимо установить, насколько
высокие показатели ICD связаны с реальными жизненными достижениями
участников после окончания вуза. Если ICD действительно отражает важные
качества – целеустремлённость, разносторонность, способность влиять на
других, – то должна наблюдаться положительная корреляция с такими
объективными метриками успеха, как поступление в магистратуру,
трудоустройство по специальности, уровень дохода, карьерный рост,
удовлетворённость жизнью.

Цель исследования: количественно оценить связь между ICD (и его
компонентами) участников движения «Мы и Наш Дома» и их последующими
жизненными успехами.

4.2. Определение и операционализация «жизненного успеха»
--------------------------------------------------------

«Жизненный успех» – многомерное понятие. Для исследования необходимо
выбрать измеримые показатели, которые можно получить с достаточной
достоверностью. Предлагается следующий набор индикаторов:

| Категория              | Показатель                                                                        | Единица измерения                                        | Источник данных                      |
|------------------------|-----------------------------------------------------------------------------------|----------------------------------------------------------|--------------------------------------|
| **Образование**        | Поступление в магистратуру / аспирантуру                                          | Бинарно (да/нет), престижность вуза (рейтинг)            | Анкетирование, данные вузов          |
|                        | Средний балл диплома (при наличии)                                                | Число (0–5 или 0–100)                                    | Анкетирование                        |
| **Трудоустройство**    | Трудоустройство по специальности                                                  | Бинарно (да/нет)                                         | Анкетирование                        |
|                        | Время до первого трудоустройства после выпуска                                    | Месяцы                                                   | Анкетирование                        |
|                        | Уровень дохода (зарплата) через 1, 3, 5 лет после выпуска                         | Рубли (в пересчёте на паритет покупательной способности) | Анкетирование (с учётом анонимности) |
| **Карьера**            | Занимаемая должность (руководитель / специалист / индивидуальный предприниматель) | Категориальная                                           | Анкетирование                        |

| **Социальный капитал** | Участие в профессиональных сообществах                                            | Бинарно                                                  | Анкетирование                        |
|                        | Публичная активность (публикации, выступления)                                    | Число                                                    | Анкетирование, анализ профилей       |
| **Удовлетворённость**  | Субъективная удовлетворённость жизнью (по шкале 1–10)                             | Число                                                    | Анкетирование                        |

4.3. Методология исследования
-----------------------------

4.3.1. Дизайн исследования
--------------------------

Продольное (лонгитюдное) исследование с ретроспективным сбором данных
для первых выпускников и проспективным наблюдением за новыми когортами.

4.3.2. Выборка
--------------

-   **Целевая группа:** участники движения, окончившие вуз не менее 1
    года назад (на момент сбора данных). Планируется охватить все
    выпуски, начиная с первого года существования движения.

-   **Контрольная группа:** студенты того же вуза, не участвовавшие в
    движении, но сопоставимые по специальности, полу, возрасту и
    успеваемости. Формируется методом подбора (matching) для устранения
    эффекта самоотбора.

-   **Размер выборки:** для обнаружения эффекта средней величины (r \~
    0.2) с мощностью 0.8 и уровнем значимости 0.05 необходимо около
    200–300 участников в каждой группе.

4.3.3. Сбор данных
------------------

1.  **Извлечение показателей ICD из цифровой системы движения** (с
    согласия участников). Фиксируются:

    -   Индивидуальный ICD на момент окончания вуза.

    -   Индекс влияния на момент окончания вуза.

    -   Значения по отдельным лучам (E, W, J, S, C).

    -   Динамика ICD за время обучения (тренд).

2.  **Анкетирование выпускников** (онлайн-опрос) для сбора данных о
    жизненных успехах. Опрос анонимный, но с возможностью связать с
    профилем через уникальный код.

3.  **Парсинг открытых источников** (LinkedIn, eLibrary, РИНЦ) для
    верификации некоторых показателей (публикации, место работы).

4.  **Запрос данных из вузов** (при наличии согласия) о среднем балле
    диплома, поступлении в магистратуру.

4.3.4. Статистический анализ
----------------------------

-   **Корреляционный анализ** Пирсона или Спирмена между
    ICD/компонентами и количественными показателями успеха (доход, балл
    диплома и т.п.).

-   **Сравнение средних** (t-критерий) между группой участников и
    контрольной группой по каждому показателю успеха.

-   **Регрессионный анализ** (линейная или логистическая регрессия) для
    выявления влияния ICD на успех с контролем возможных смешивающих
    факторов (специальность, пол, исходная успеваемость,
    социально-экономический статус семьи).

-   **Анализ выживаемости** (Cox-регрессия) для времени до
    трудоустройства.

4.3.5. Этические аспекты
------------------------

-   Получение информированного согласия от участников.

-   Анонимизация данных при публикации результатов.

-   Соблюдение законодательства о персональных данных.

4.4. Гипотезы исследования
--------------------------

На основе теоретических предпосылок можно сформулировать следующие
гипотезы:

1.  Участники с более высоким ICD (особенно сбалансированным) чаще
    поступают в магистратуру и аспирантуру, чем их сверстники с низким
    ICD.

2.  Индекс влияния положительно коррелирует с занятием руководящих
    должностей и уровнем дохода через 5 лет после выпуска.

3.  Высокие значения по лучу C (культура) связаны с научной и творческой
    карьерой (публикации, патенты).

4.  Высокие значения по лучу W (благо) и J (справедливость) коррелируют
    с работой в некоммерческом секторе, государственной службе.

5.  Динамика ICD (рост за время обучения) предсказывает успех лучше, чем
    конечное значение (показывает способность к развитию).

4.5. Возможные проблемы и способы их решения
--------------------------------------------

| Проблема                                                                             | Способ решения                                                                                                          |
|--------------------------------------------------------------------------------------|-------------------------------------------------------------------------------------------------------------------------|
| **Самоотбор** (более мотивированные студенты идут в движение, они же и так успешнее) | Использование контрольной группы с matching; учёт в регрессии исходных характеристик (ЕГЭ, средний балл до вступления). |
| **Временной лаг** (первые выпускники появились недавно, данных мало)                 | Начать исследование проспективно, параллельно собирая ретроспективные данные по мере накопления.                        |
| **Неполнота данных** (не все участники отвечают на опросы)                           | Стимулирование участия (бонусные баллы, розыгрыши), использование открытых источников, взвешивание ответов.             |
| **Субъективность самоотчётов о доходе**                                              | Запрос диапазона вместо точной суммы, верификация через LinkedIn/профессиональные сети.                                 |
| **Конфиденциальность**                                                               | Шифрование данных, разделение ролей (администратор базы не видит ответы).                                               |

4.6. Ожидаемые результаты и их интерпретация
--------------------------------------------

Если гипотезы подтвердятся, это будет сильным аргументом в пользу того,
что ICD действительно измеряет значимые для жизни качества. Конкретные
количественные оценки (например, «увеличение ICD на 0,1 связано с ростом
зарплаты на 15%») позволят использовать ICD как прогностический
инструмент.

Напротив, отсутствие корреляции может указывать на необходимость
корректировки методики расчёта ICD, пересмотра лучей или весов. Даже
нулевой результат будет полезен для понимания ограничений индекса.

4.7. Практическая значимость
----------------------------

1.  **Валидация индекса** – подтверждение, что ICD действительно
    отражает потенциал успешности.

2.  **Улучшение профориентации** – зная, какие лучи сильнее коррелируют
    с определёнными карьерными траекториями, можно давать рекомендации
    участникам.

3.  **Обоснование для работодателей** – демонстрация, что выпускники с
    высоким ICD обладают ценными качествами, что повысит спрос на них.

4.  **Совершенствование системы** – выявление слабых мест методики,
    корректировка порогов статусов, весов лучей.

4.8. Заключение
---------------

Исследование корреляции ICD с реальными жизненными успехами – ключевой
этап в развитии научной базы движения «Мы и Наш Дома». Оно позволит не
только подтвердить практическую значимость индекса, но и получить данные
для его дальнейшего улучшения. Результаты будут публиковаться в открытых
отчётах и использоваться для привлечения новых участников, партнёров и
спонсоров.

5. Анализ оптимальных порогов для статусов (Мастер, Творец, Легенда) на основе статистики движения
--------------------------------------------------------------------------------------------------

5.1. Постановка задачи
----------------------

В системе мотивации движения «Мы и Наш Дома» предусмотрены три статуса,
присваиваемые автоматически при достижении определённых
значений **индекса влияния** (Impact Index):

-   **Мастер** – порог $P_{M}$;

-   **Творец** – порог $P_{T}$;

-   **Легенда** – порог $P_{L}$.

Первоначально были установлены ориентировочные
значения $P_{M} = 1000$, $P_{T} = 5000$, $P_{L} = 10000$. Однако эти
цифры являются экспертными и не опираются на реальное распределение
индекса влияния среди участников. По мере накопления данных возникает
необходимость в **научно обоснованном выборе порогов**, которые бы:

-   отражали естественные разрывы в распределении достижений;

-   были достаточно высокими, чтобы статус сохранял престиж и редкость;

-   были достижимыми для мотивации активных участников;

-   учитывали динамику роста движения и возможные изменения в
    распределении со временем.

5.2. Теоретические подходы к определению порогов
------------------------------------------------

5.2.1. Квантильный подход
-------------------------

Один из распространённых способов – установить пороги как определённые
процентили распределения индекса влияния. Например:

-   Мастер – 80-й процентиль (20% лучших);

-   Творец – 95-й процентиль (5% лучших);

-   Легенда – 99-й процентиль (1% лучших).

Это гарантирует, что статусы будут присваиваться фиксированной доле
участников независимо от абсолютных значений, что удобно для сравнения
во времени и между разными по размеру сообществами. Однако такой подход
не учитывает возможные естественные разрывы и может быть чувствителен к
выбросам.

5.2.2. Метод естественных разрывов (Jenks natural breaks)
---------------------------------------------------------

Используется в картографии и статистике для группировки данных так,
чтобы минимизировать внутригрупповую дисперсию и максимизировать
межгрупповую. Применительно к нашему случаю можно искать точки, где
распределение индекса влияния естественным образом разделяется на
кластеры (например, «обычные участники», «активисты», «лидеры»,
«легенды»). Этот метод объективнее, но требует достаточно большого
объёма данных и может давать разные результаты при добавлении новых
наблюдений.

5.2.3. Целевой подход, исходя из миссии движения
------------------------------------------------

Статусы должны отражать не просто статистические группы, а качественные
уровни вклада. Можно определить:

-   **Мастер** – участник, чей вклад уже заметно выделяется на общем
    фоне, он может быть наставником для новичков.

-   **Творец** – участник, создавший значимые проекты, повлиявшие на
    многих.

-   **Легенда** – исключительный вклад, сопоставимый с прорывными
    достижениями.

В этом случае пороги могут быть установлены на основе экспертных оценок
и затем калиброваться по мере накопления данных о том, какие значения
индекса влияния реально соответствуют таким качественным описаниям.

5.2.4. Комбинированный подход
-----------------------------

На начальном этапе, пока данных мало, используются экспертные пороги.
Затем, по мере накопления статистики, они корректируются на основе
анализа распределения (например, с использованием квантилей или метода
естественных разрывов). При этом важно сохранить преемственность: старые
статусы не должны быть «девальвированы» резким изменением порогов.

5.3. Методология определения оптимальных порогов
------------------------------------------------

5.3.1. Сбор данных
------------------

Для анализа необходимы значения индекса влияния всех участников
движения, начиная с самого первого. Данные извлекаются из системы учёта
(токены KIT или таблицы). Важно исключить неактивных участников
(например, тех, у кого нет деяний более года), чтобы распределение
отражало реальную активность.

5.3.2. Статистический анализ
----------------------------

1.  **Построение гистограммы распределения** индекса влияния. Обычно
    распределение таких показателей (сумма взвешенных баллов) имеет вид,
    близкий к логнормальному или с «тяжёлым хвостом».

2.  **Вычисление основных дескриптивных статистик**: среднее, медиана,
    квантили (90%, 95%, 99%, 99.9%).

3.  **Применение метода естественных разрывов** (алгоритм Дженкса) для
    поиска оптимальных точек разделения на 3 или 4 группы. Для этого
    можно использовать библиотеки Python (jenkspy) или R (BAMMtools).

4.  **Анализ динамики** – как меняются квантили со временем. Если
    движение быстро растёт, квантили могут смещаться, поэтому пороги,
    основанные на процентилях, могут быть более устойчивы.

5.3.3. Учёт качественных критериев
----------------------------------

На основе анализа можно предложить несколько вариантов порогов. Затем
они обсуждаются с Советом движения и, возможно, с фокус-группами из
действующих Мастеров и Творцов, чтобы проверить, соответствуют ли эти
пороги интуитивному представлению о статусах.

5.3.4. Механизм периодического пересмотра
-----------------------------------------

Поскольку движение развивается, пороги не должны быть застывшими.
Предлагается следующий регламент:

-   Раз в 2 года (или при удвоении числа участников) проводится анализ
    распределения.

-   Пороги корректируются, но с условием, что уже присвоенные статусы
    сохраняются (дедушка-оговорка). Новые участники получают статусы по
    новым правилам.

-   Коррекция не должна быть слишком резкой; допустимо плавное изменение
    (например, усреднение со старыми значениями).

5.4. Иллюстративный пример (гипотетические данные)
--------------------------------------------------

Предположим, что на текущий момент в движении 10 000 участников, у
которых индекс влияния распределён следующим образом:

| Процентиль    | Значение индекса влияния |
|---------------|--------------------------|
| 50% (медиана) | 120                      |
| 80%           | 450                      |
| 90%           | 1100                     |
| 95%           | 2800                     |
| 99%           | 15000                    |
| 99.9%         | 80000                    |

Анализ естественных разрывов может показать точки около 500, 3000 и
20000. Тогда:

-   Мастер – от 500 до 3000 (попадают \~15% участников);

-   Творец – от 3000 до 20000 (\~4%);

-   Легенда – более 20000 (\~1%).

Сравним с предварительными порогами (1000, 5000, 10000). Видно, что
порог Мастера 1000 был бы чуть выше 90-го процентиля, что делает статус
слишком редким (только 10% участников). Возможно, стоит снизить порог
Мастера до 500, чтобы охватить примерно 20% наиболее активных. Порог
Творца 5000 попадает между 95% и 99% – около 3–4%, что разумно. Порог
Легенды 10000 – около 98,5-го процентиля, но по естественным разрывам
видна точка 20000, которая может лучше соответствовать «легендарности».
Возможно, стоит повысить порог Легенды до 20000, чтобы он действительно
отражал исключительные заслуги.

5.5. Учёт ICD в присвоении статусов
-----------------------------------

В текущей модели статусы определяются только по индексу влияния, что
стимулирует социальную активность. Однако гармоничное развитие (ICD)
тоже важно. Можно рассмотреть комбинированный критерий, например:

-   **Мастер**: индекс влияния \> $P_{M}$ **или** ICD \> 0,7
    (гармоничная личность).

-   **Творец**: индекс влияния \> $P_{T}$ **и** ICD \> 0,5 (сочетание
    влияния и гармонии).

-   **Легенда**: индекс влияния \> $P_{L}$ **и** ICD \>
    0,6 **и** наличие прорывного деяния.

Такой подход потребует дополнительного анализа, но может лучше отражать
философию «Мы и Наш Дом».

5.6. Рекомендации по внедрению
------------------------------

1.  **Накопить данные** как минимум за 2 года активной работы движения
    (чтобы распределение стало устойчивым).

2.  **Провести предварительный анализ** на пилотных данных (первый вуз)
    и скорректировать пороги до масштабирования.

3.  **Утвердить методику пересмотра** (например, раз в 2 года) и
    зафиксировать её в нормативных документах.

4.  **Информировать участников** о правилах присвоения статусов и
    возможной корректировке порогов, чтобы избежать недовольства.

5.  **Рассмотреть введение дополнительных градаций** (например, Мастер
    1-й степени, Мастер 2-й степени) для более плавной мотивации, если
    распределение окажется сильно растянутым.

5.7. Заключение
---------------

Оптимальные пороги статусов должны определяться на основе
статистического анализа распределения индекса влияния, с учётом целей
движения и качественных представлений о значимости статусов.
Предложенный подход (квантили + метод естественных разрывов) позволяет
объективно и прозрачно устанавливать пороги, а механизм периодического
пересмотра обеспечивает их актуальность по мере роста движения.
Дальнейшие исследования могут включить в критерии также ICD и другие
показатели, отражающие гармоничность развития.

ЗАКЛЮЧИТЕЛЬНАЯ ЧАСТЬ. ЭКОНОМИЧЕСКИЕ И ФИНАНСОВЫЕ АСПЕКТЫ
========================================================

1. Механизмы привлечения инвестиций на каждом уровне TRL
--------------------------------------------------------

Устойчивое развитие движения «Мы и Наш Дом» требует диверсифицированной
модели финансирования, учитывающей специфику каждого этапа зрелости
проекта. Ниже представлена поэтапная стратегия привлечения ресурсов,
основанная на анализе существующих в России механизмов поддержки
молодёжных и социальных проектов.

1.1 Общая структура финансирования движения
-------------------------------------------

| Этап       | Основные источники                                                                  | Целевое использование                                             | Ожидаемый объём |
|------------|-------------------------------------------------------------------------------------|-------------------------------------------------------------------|-----------------|
| **TRL1–2** | Собственные средства инициаторов, краудфандинг, микрогранты                         | Разработка концепции, создание прототипов, юридическое оформление | 0–300 тыс. руб. |
| **TRL3**   | Гранты физическим лицам (Росмолодёжь), краудфандинг, спонсорство                    | Пилотные проекты, развитие цифровой платформы, мероприятия        | 0,5–2 млн руб.  |
| **TRL4**   | Гранты НКО, корпоративные партнёрства, региональные субсидии                        | Масштабирование в одном вузе, создание инфраструктуры             | 2–5 млн руб.    |
| **TRL5**   | Президентские гранты, фонды, бизнес-партнёрства, эндаумент                          | Региональное развитие, обучение актива, проведение слётов         | 5–15 млн руб.   |
| **TRL6**   | Государственно-частное партнёрство, международные гранты, коммерциализация сервисов | Федеральный охват, цифровая платформа, международное развитие     | 15–50+ млн руб. |

1.2. Детализация механизмов по этапам
-------------------------------------

1.2.1. TRL1–2: Идея и концепция
-------------------------------

**Ключевые источники:**

1.  **Собственные средства инициативной группы** – базовый ресурс для
    запуска.

2.  **Микрогранты и конкурсы для студенческих инициатив** – многие вузы
    имеют внутренние фонды поддержки студенческих проектов (до 100–200
    тыс. руб.) .

3.  **Краудфандинг (модель вознаграждения)** – на этом этапе идеально
    подходит для проверки гипотезы и сбора первого сообщества.
    Вознаграждениями могут быть: упоминание на сайте, значок
    «Первопроходец», доступ к закрытому чату .

**Обоснование:** На ранних стадиях венчурные инвесторы и крупные фонды,
как правило, не готовы рассматривать проекты без подтверждённой
эффективности. Эксперты по социальным стартапам подчёркивают: «Венчурные
фонды – это (обычно) неправильное место для начала» . Краудфандинг же
позволяет не только собрать средства, но и сформировать лояльное
сообщество, что критически важно для движения .

**Примеры из практики:** Проект «Театр Села» в Калужской области начал с
краудфандинга и поэтапно собрал более миллиона рублей на выкуп
помещения, ремонт и развитие . Социальные и благотворительные инициативы
в 2023 году собрали на [Planeta.ru](https://planeta.ru/) 65 млн рублей
(+20% к предыдущему году) .

1.2.2. TRL3: Эксперимент
------------------------

**Ключевые источники:**

1.  **Гранты «Росмолодёжь.Гранты»** – основной инструмент поддержки
    молодёжных инициатив в России. Грантополучателями могут выступать
    физические лица в возрасте 14–35 лет, сумма до 1 млн рублей на
    карту. Номинации: \#вдохновляй (творческие проекты), \#экологика
    (экологические), \#создавай\_возможности (социальные лифты) и др. .

2.  **Программа «УМНИК» Фонда содействия инновациям** – для студентов и
    аспирантов до 35 лет, имеющих научную идею. Грант 500 тыс. рублей на
    коммерциализацию .

3.  **Конкурс «Студенческий стартап»** – до 1 млн рублей на развитие
    бизнес-идеи .

4.  **Краудфандинг (продолжение)** – на этом этапе можно привлекать
    средства на конкретные пилотные мероприятия (например, проведение
    первого слёта).

**Обоснование:** На стадии эксперимента проект уже имеет первые
результаты, что повышает доверие. Грантовые конкурсы, особенно
«Росмолодёжь.Гранты», ориентированы именно на социальные проекты и не
требуют регистрации юридического лица . Важное условие – чёткая
социальная направленность и измеримый результат . Движение «Мы и Наш
Дом» с его пятью лучами полностью соответствует этим требованиям.

1.2.3. TRL4: Пилотный вуз
-------------------------

**Ключевые источники:**

1.  **Гранты для некоммерческих организаций** – на этом этапе
    рекомендуется зарегистрировать юридическое лицо (например, АНО или
    общественную организацию) для доступа к более крупным грантам:

    -   **Фонд президентских грантов** – один из крупнейших источников
        финансирования социальных проектов в России. Гранты получают
        тысячи НКО ежегодно .

    -   **Гранты в форме субсидий** – согласно федеральному бюджету,
        предусмотрены субсидии молодёжным и экологическим организациям
        (например, Всероссийскому экологическому движению «Экосистема» –
        2,78 млрд руб. в 2025 году) . Хотя такие суммы доступны крупным
        организациям, наличие успешного пилотного проекта позволяет
        претендовать на региональные субсидии.

2.  **Корпоративные партнёрства с локальным бизнесом** – компании
    заинтересованы в подготовке кадров и развитии территорий
    присутствия. Например, благотворительный фонд АФК «Система» активно
    сотрудничает с региональными университетами, открывая лаборатории,
    поддерживая стажировки и профориентационные проекты .

3.  **Поддержка администрации вуза** – выделение помещений,
    софинансирование мероприятий, помощь в организации.

4.  **Региональные субсидии** – многие субъекты РФ имеют программы
    поддержки молодёжных и социальных проектов.

**Обоснование:** На этапе пилотного вуза движение демонстрирует
устойчивость и измеримые результаты, что делает его привлекательным для
грантодателей и бизнес-партнёров. Опыт АФК «Система» показывает, что
корпорации готовы инвестировать в создание лабораторий, проведение
стажировок и конкурсов, если видят заинтересованность студентов и связь
с реальным сектором экономики .

1.2.4. TRL5: Региональный рост
------------------------------

**Ключевые источники:**

1.  **Крупные гранты федеральных фондов** – Фонд президентских грантов,
    гранты Минобрнауки, Минпросвещения.

2.  **Венчурная филантропия** – подход, при котором инвесторы
    предоставляют не только финансирование, но и управленческую
    экспертизу, на долгосрочной основе (3–7 лет). Характерные черты:
    высокая вовлечённость инвестора, ориентация на социальный эффект,
    готовность экспериментировать .

3.  **Корпоративные партнёрства на федеральном уровне** – системное
    взаимодействие с крупными компаниями, заинтересованными в развитии
    кадрового потенциала и корпоративной социальной ответственности.

4.  **Эндаумент (целевой капитал)** – начало формирования фонда целевого
    капитала из пожертвований выпускников и партнёров для обеспечения
    долгосрочной устойчивости.

5.  **Региональные слёты и мероприятия** – частичное самофинансирование
    за счёт регистрационных взносов (символических для участников) и
    спонсорских пакетов.

**Обоснование:** На этапе регионального роста движение охватывает тысячи
участников, что создаёт основу для привлечения серьёзных инвестиций.
Венчурная филантропия особенно релевантна, так как движение имеет чётко
измеримые социальные результаты (ICD, индекс влияния, количество
вовлечённых), что позволяет оценивать эффективность вложений .

1.2.5. TRL6: Всероссийский масштаб
----------------------------------

**Ключевые источники:**

1.  **Государственно-частное партнёрство** – реализация масштабных
    программ при поддержке государства и бизнеса (например, создание
    федеральной цифровой платформы, проведение всероссийских форумов).

2.  **Международные гранты** – программы молодёжных обменов, совместные
    проекты с зарубежными фондами (при выходе на международный уровень).

3.  **Коммерциализация сервисов** – разработка и внедрение цифровых
    продуктов на базе системы ICD для корпоративных клиентов (HR-оценка,
    развитие soft skills), проведение платных образовательных программ,
    консалтинг.

4.  **Фандрайзинг среди выпускников** – создание системы регулярных
    пожертвований от успешных выпускников, которые получили пользу от
    участия в движении.

5.  **Продажа мерча** – брендированная продукция как дополнительный
    источник дохода и инструмент продвижения.

**Обоснование:** На федеральном уровне движение становится узнаваемым
брендом, что открывает возможности для коммерциализации отдельных
направлений без потери социальной миссии. Опыт крупнейших молодежных
организаций (например, «Россия – страна возможностей») показывает, что
сочетание грантового финансирования, партнёрств и собственных доходов
обеспечивает устойчивость.

1.3. Ключевые принципы привлечения инвестиций
---------------------------------------------

1.  **Прозрачность и измеримость результатов** – наличие чётких метрик
    (ICD, индекс влияния, количество участников, охват) повышает доверие
    инвесторов. Проекты с активной коммуникацией и прозрачной
    отчётностью собирают в 2–3 раза больше средств .

2.  **Диверсификация источников** – сочетание грантов, партнёрств,
    краудфандинга и собственных доходов снижает риски.

3.  **Построение сообщества вокруг миссии** – лояльные участники
    становятся не только волонтёрами, но и донорами, амбассадорами.
    Краудфандинг на ранних этапах помогает сформировать это сообщество .

4.  **Фокус на социальном воздействии** – все инвестиции должны быть
    направлены на достижение целей движения, а не на обогащение. Это
    соответствует принципам венчурной филантропии .

5.  **Партнёрства с сильными игроками** – поиск партнёров, которые
    сильнее вас в нужных областях (финансы, технологии, связи), делает
    движение устойчивее .

2. Анализ рисков перехода с Polygon на собственную блокчейн-систему
-------------------------------------------------------------------

**Введение**

В рамках развития движения «Мы и Наш Дома» и системы ICD рассматривается
вопрос о долгосрочной стратегии использования блокчейн-инфраструктуры. В
настоящее время проект базируется на сети Polygon (публичный
PoS-блокчейн, L2 для Ethereum). Однако по мере роста движения может
возникнуть соблазн создать **собственную блокчейн-систему** (как это
сделали Polymarket, крупные DeFi-протоколы). Данный анализ оценивает
риски и целесообразность такого перехода, опираясь на реальные кейсы и
экспертные оценки.

2.1. Мотивация для перехода на собственную сеть
-----------------------------------------------

2.1.1. Аргументы «ЗА» (почему проекты уходят)
---------------------------------------------

**Кейс Polymarket (декабрь 2025)** :

-   **Причина ухода:** нестабильность сети Polygon (15 сбоев, инцидентов
    или простоев в 2025 году, включая 24-часовую задержку транзакций
    12–13 декабря, что напрямую влияло на исполнение ордеров).

-   **Экономическая мотивация:** Polymarket генерировал \~25% всего
    объёма транзакций и газа в сети Polygon. Создание собственной L2
    позволяет «замкнуть» экономическую активность внутри своей
    экосистемы.

-   **Контроль над оракулами:** Проблемы с внешними оракулами (UMA)
    приводили к спорным исходам на миллионы долларов. Собственная сеть
    даёт возможность создать нативные оракулы на стейкинге токена POLY.

**Потенциальные выгоды для «Мы и Наш Дом»:**

-   Полный контроль над параметрами сети (блоки, комиссии,
    конфиденциальность данных участников).

-   Возможность встроить механизмы расчёта ICD непосредственно в
    протокол (нативные смарт-контракты).

-   Создание собственной экономики токена (стейкинг, комиссии,
    управление).

-   Повышение устойчивости к цензуре и независимость от внешних сбоев.

2.1.2. Аргументы «ПРОТИВ» (почему это преждевременно)
-----------------------------------------------------

**Анализ текущей ситуации движения «Мы и Наш Дома»:**

-   **Масштаб:** Движение находится на ранних этапах (TRL1–3). Даже в
    пилотном вузе речь идёт о сотнях, а не миллионах пользователей.

-   **Техническая сложность:** Создание и поддержка собственного
    блокчейна требует команды квалифицированных разработчиков (как
    минимум 3–5 человек на полную ставку), что недоступно на начальных
    этапах.

-   **Безопасность:** Собственная сеть с небольшим числом валидаторов
    будет уязвима для атак (например, 51%-атака), в то время как Polygon
    наследует безопасность Ethereum .

2.2. Сравнение с использованием готовой инфраструктуры Polygon
--------------------------------------------------------------

Текущие затраты на использование Polygon :

-   Транзакционные комиссии: минимальны (доли цента).

-   Хеширование таблиц: одна транзакция в месяц (\~\$0.01–\$0.05).

-   Интеграция: бесплатные инструменты (Remix, MetaMask, OpenZeppelin).

-   Безопасность: наследуется от Ethereum, сеть с многомиллиардной
    капитализацией.

**Экономия:** порядка \$50,000–\$150,000 в год по сравнению с
разработкой собственной сети.

2.2.3. Риски, связанные с Polygon
---------------------------------

**Нестабильность сети:** Как показал кейс Polymarket, Polygon может
испытывать сбои . Для движения «Мы и Наш Дома», где запись хешей
происходит раз в месяц, кратковременные сбои не критичны. Однако если в
будущем потребуется более частое взаимодействие с блокчейном (например,
для микроплатежей), этот риск возрастёт.

**Риск централизации:** Polygon критикуют за чрезмерную централизацию
управления , что противоречит философии децентрализации. Однако для
студенческого движения, где доверие к организаторам и так высоко, этот
риск не является приоритетным.

2.3. Оценка необходимости для разных сценариев развития
-------------------------------------------------------

2.3.1. Сценарий А (ближайшие 3–5 лет)
-------------------------------------

-   **Потребности:** ежемесячная запись хешей, голосование, базовые
    токены KIT.

-   **Достаточно:** Polygon (или другой публичный L2). Все функции
    реализуются через смарт-контракты на существующей сети.

-   **Риски перехода:** неоправданные затраты, отвлечение ресурсов от
    развития сообщества, потеря времени на технические проблемы.

2.3.2. Сценарий Б (5–10 лет, при росте до 100 000+ участников)
--------------------------------------------------------------

-   **Потребности:** высокая частота транзакций, сложные механизмы
    голосования, возможная монетизация.

-   **Альтернатива:** использование **Polygon CDK (Chain Development
    Kit)** . Это позволяет запустить собственную L2-сеть, совместимую с
    Ethereum и Polygon, но с гораздо меньшими затратами, чем создание L1
    с нуля. CDK предоставляет готовые модули (консенсус, виртуальная
    машина, мосты), сокращая разработку на 30–40% .

2.3.3. Сценарий В (международный масштаб, миллионы участников)
--------------------------------------------------------------

-   **Потребности:** полный суверенитет, собственная экономика токена,
    устойчивость к цензуре.

-   **Необходимость:** создание собственной L1 или суверенного L2 с
    собственными валидаторами.

-   **Риски:** огромные затраты, необходимость поддержания безопасности,
    сложность привлечения валидаторов.

2.5. Выводы и рекомендации для движения «Мы и Наш Дома»
-------------------------------------------------------

На основе проведённого анализа можно сделать следующие выводы:

1.  **На этапах TRL1–TRL5 (ближайшие 3–5 лет) переход на собственную
    сеть нецелесообразен и крайне рискован.** Основные причины:

    -   Непропорционально высокие затраты (в 1000+ раз выше текущих).

    -   Отсутствие критической необходимости (текущие функции успешно
        реализуются на Polygon).

    -   Риск потери фокуса на развитии сообщества из-за технических
        сложностей.

2.  **Polygon остаётся оптимальным выбором для движения** благодаря:

    -   Нулевым затратам на инфраструктуру.

    -   Высокой безопасности (наследование от Ethereum) .

    -   Широкой поддержке инструментов и комьюнити.

    -   Успешному опыту крупных проектов (VK, Starbucks, Disney) .

3.  **Мониторинг ситуации:** необходимо отслеживать стабильность сети
    Polygon. В 2025 году зафиксировано 15 сбоев , что может стать
    проблемой при росте активности. Однако для ежемесячного хеширования
    это не критично.

4.  **Перспективная стратегия:** при достижении масштаба в десятки тысяч
    активных участников и появлении финансовых ресурсов рассмотреть
    возможность перехода на **Polygon CDK** для запуска собственной L2.
    Это позволит:

    -   Сохранить совместимость с Ethereum.

    -   Получить контроль над параметрами сети.

    -   Существенно снизить затраты по сравнению с L1.

    -   Использовать готовые модули и инструменты.

5.  **Критическое предостережение:** Никогда не следует экономить на
    аудите безопасности . Уязвимость смарт-контракта может стоить
    репутации движения, которая нарабатывается годами.

Заключение
----------

Переход с Polygon на собственную блокчейн-инфраструктуру для движения
«Мы и Наш Дома» на текущем этапе развития
представляется **преждевременным и экономически неоправданным**. Гораздо
важнее сконцентрироваться на развитии сообщества, методической базы и
привлечении участников. Использование готовой, надёжной и бесплатной
инфраструктуры Polygon позволяет минимизировать технические риски и
направить все ресурсы на достижение главной цели — перехода к УЦ-2.

В долгосрочной перспективе (5–10 лет) при масштабировании до миллионов
участников и появлении устойчивого финансирования может быть рассмотрен
вариант создания собственной L2 на базе Polygon CDK, что обеспечит
баланс между контролем и затратами.

3.3. Модель монетизации и возврат инвестиций
